```python
tamil_text = """<doc id="1206" title="திருச்சிராப்பள்ளி">
{{dablink|இக்கட்டுரை நகரம் பற்றியது, இதே பெயரில் உள்ள மாவட்டம் பற்றி அறிய [[திருச்சிராப்பள்ளி மாவட்டம்]] மற்றும் [[திருச்சிராப்பள்ளி மாநகராட்சி]] கட்டுரையைப் பார்க்க.}}
 
 {{Infobox settlement
 | name = திருச்சிராப்பள்ளி
 | native_name = 
 | other_name = திருச்சி
 | settlement_type = [[மாநகரம்]]
 | image_skyline = Trichy montage1.jpg
 | image_alt = 
 | image_caption = மேலிருந்து கடிகாரம் சுழலும் திசையில்:[[திருவானைக்காவல் ஜம்புகேசுவரர் கோயில்|ஜம்புகேசுவரர் கோயில்]], [[திருவரங்கம் அரங்கநாத சுவாமி கோயில்|அரங்கநாதசுவாமி கோயில்]], தூய லூர்து அன்னை கிறித்தவத் தேவாலயம், [[திருச்சி மலைக் கோட்டை|மலைக்கோட்டை]], திருச்சிராப்பள்ளியை [[திருவரங்கம் தீவு|ஸ்ரீரங்கம் தீவிலிருந்து]] பிரிக்கும் [[காவிரி ஆறு]], [[மேலணை]]
 | nickname = மலைக்கோட்டை மாநகரம்
 | map_alt = 
 | map_caption = 
 | pushpin_map = India Tamil Nadu#India
 | pushpin_label_position = left
 | pushpin_map_alt = 
 | pushpin_map_caption = திருச்சிராப்பள்ளி, தமிழ்நாடு
 | coordinates = {{coord|10|47|25|N|78|42|17|E|display=inline,title}}
 | subdivision_type = நாடு
 | subdivision_name = {{flag|இந்தியா}}
 | subdivision_type1 = [[இந்தியாவின் மாநிலங்களும் ஆட்சிப்பகுதிகளும்|மாநிலம்]]
 | subdivision_type2 = [[மாவட்டம் (இந்தியா)|மாவட்டம்]]
 | subdivision_name1 = [[தமிழ்நாடு]]
 | subdivision_name2 = [[திருச்சிராப்பள்ளி மாவட்டம்|திருச்சிராப்பள்ளி]]
 | established_title = &lt;!-- Established --&gt;
 | established_date = 1866
 | founder = 
 | named_for = 
 | parts_type = பகுதி
 | parts = [[சோழ நாடு]]
 | government_type = மாநகராட்சி
 | governing_body = [[திருச்சிராப்பள்ளி மாநகராட்சி]]
 | leader_title1 = [[மக்களவை (இந்தியா)|மக்களவை உறுப்பினர்]]
 | leader_name1 = [[சு. திருநாவுக்கரசர்]]
 | leader_title2 = [[தமிழ்நாடு சட்டப் பேரவை|சட்டமன்ற உறுப்பினர்]]
 | leader_name2 = [[இனிகோ எஸ். இருதயராஜ்]] {{small|([[திருச்சிராப்பள்ளி கிழக்கு (சட்டமன்றத் தொகுதி)|திருச்சி கிழக்கு]])}} &lt;br /&gt; [[கே. என். நேரு]] {{small|([[திருச்சிராப்பள்ளி மேற்கு (சட்டமன்றத் தொகுதி)|திருச்சி மேற்கு]])}}&lt;br /&gt; [[அன்பில் மகேஷ் பொய்யாமொழி]] {{small|([[திருவெறும்பூர் (சட்டமன்றத் தொகுதி)|திருவெறும்பூர்]])}} &lt;br /&gt; [[மொ. பழனியாண்டி]] {{small|([[திருவரங்கம் (சட்டமன்றத் தொகுதி)|திருவரங்கம்]])}}
 | leader_title3 = [[திருச்சிராப்பள்ளி மாநகராட்சி|மாநகர மேயர்]]
 | leader_name3 = திரு. அன்பழகன் [[திமுக]]
 | leader_title4 = மாவட்ட ஆட்சியர்
 | leader_name4 = பிரதீப் குமார்
 
 , [[இந்திய ஆட்சிப் பணி|இ.ஆ.ப]]
 | unit_pref = Metric
 | area_footnotes = 
 | area_total_km2 = 167.23
 | area_metro_km2 = 211.51
 | area_rank = 3
 | elevation_m = 105
 | population_total = 847387
 | population_as_of = 2011
 | population_footnotes = 
 | population_density_km2 = auto
 | population_metro = 1221717
 | population_metro_footnotes = &lt;ref name=&quot;UA_2011&quot;&gt;{{cite web|title=Primary Census Abstract - Urban Agglomeration |url=http://www.censusindia.gov.in/DigitalLibrary/MFTableSeries.aspx |format=XLS |publisher=Registrar General and Census Commissioner of India|accessdate=13 October 2015|url-status=dead |archiveurl=https://web.archive.org/web/20160315050316/http://www.censusindia.gov.in/DigitalLibrary/MFTableSeries.aspx |archivedate=15 March 2016 }}&lt;/ref&gt;
 | population_rank = 4
 | population_demonym = [[தமிழர்]]
 | demographics_type1 = மொழிகள்
 | demographics1_title1 = அலுவல்மொழி
 | demographics1_info1 = [[தமிழ்]]
 | timezone1 = [[இந்திய சீர் நேரம்|இசீநே]]
 | utc_offset1 = +5:30
 | postal_code_type = [[அஞ்சல் குறியீட்டு எண்|அஞ்சல் குறியீடு]]
 | postal_code = 620 0xx &amp; 621 0xx
 | area_code = 0431
 | area_code_type = [[இந்தியாவில் தொலைபேசி எண்கள்|தொலைபேசி குறியீடு]]
 | registration_plate = TN-45, TN-48, TN-81
 | blank1_name_sec1 = [[சென்னை]]யிலிருந்து தொலைவு
 | blank1_info_sec1 = 320 கி.மீ (207 மைல்)
 | blank2_name_sec1 = 
 | blank2_info_sec1 = 
 | blank3_name_sec1 = 
 | blank3_info_sec1 = 
 | website = {{URL|http://www.trichycorporation.gov.in/|Trichy City Municipal Corporation}}
 | footnotes = 
 }}
 '''திருச்சிராப்பள்ளி''' (''Tiruchirappalli'' அல்லது ''Trichinopoly''&lt;ref group=&quot;கு&quot;&gt;Trichinopoly என்று பிரித்தானிய வரலாற்று ஆவணங்களில் குறிப்பிடப்படுகிறது.&lt;/ref&gt;), [[இந்தியா]]வில் உள்ள [[தமிழ்நாடு]] [[இந்தியாவின் மாநிலங்களும் ஆட்சிப்பகுதிகளும்|மாநிலத்தின்]] மையப் பகுதியில் அமைந்துள்ள மிகவும் பழமை பெற்ற தொன்மை வாய்ந்த மாநகரமாகும். இது [[திருச்சிராப்பள்ளி மாவட்டம்|திருச்சிராப்பள்ளி]] மாவட்டத்தின் தலைநகர் ஆகும். இது தமிழ்நாட்டின் பெருநகரங்களுள் ஒன்றாகும். இந்நகரம் தமிழ்நாட்டின் பரப்பளவு அடிப்படையில் மூன்றாவது பெரிய மாநகரமும் மக்கள் தொகை அடிப்படையில் நான்காவது பெரிய மாநகரமும் ஆகும். இந்த நகரம் [[தென்னிந்தியா|தென்னிந்தியாவின்]] [[சென்னை]], [[ஹைதராபாத்]],  [[பெங்களூர்]], [[கோயம்புத்தூர்]], [[கொச்சி|கொச்சிக்கு]] அடுத்த ஆறாவது மிகப்பெரிய நகரம் ஆகும். மேலும் இது [[தமிழ்நாடு உள்ளாட்சி அமைப்புகள்|உள்ளாட்சி அமைப்பில்]] [[திருச்சிராப்பள்ளி மாநகராட்சி|மாநகராட்சி]] எனும் தகுதி பெற்றது. [[காவிரி ஆறு|காவிரி ஆற்றங்கரையில்]] அமைந்துள்ள திருச்சிராப்பள்ளி சங்க காலத்தில் [[முற்காலச் சோழர்கள்|முற்கால சோழர்]]களின் தலைநகரமாகவும், தற்போதைய தமிழகத்தின் முக்கியமான நான்காவது பெரிய நகரமாகவும் உள்ளது. இதைப் பொதுவாகத் '''திருச்சி''' (''Trichy'') என்று சுருக்கமாக அழைக்கிறார்கள். திருச்சிராப்பள்ளி என்பதன் பொருளானது, திரு - சிராய் - பள்ளி, அதாவது சிராய் (சிராய் என்பது பாறை என்று பொருள்படும்) பள்ளி கொண்ட இடம். பிரசித்தி பெற்ற [[திருச்சி மலைக் கோட்டை|மலைக் கோட்டை]] இந்தப் பாறையின் மேல் பகுதியில் இந்தியாவில் எவ்விடத்திலும் இல்லாத [[இந்து மதம்|இந்து சமய]]த்தின் முழு முதற் கடவுளான [[விநாயகர்|விநாயகப் பெருமானுக்கு]] உச்சிப் பிள்ளையார் என்ற பெயரில் தனிக்கோயில் அமைந்துள்ளது. [[இந்தியா]]வின் தூய்மையான 10 நகரங்களில், [[திருச்சி]]யும் ஒன்றாகும். முன்னாள் தமிழக முதலமைச்சர் [[எம்.ஜி.ராமச்சந்திரன்]] அவர்கள் (1980–1984) ஆட்சி காலத்தில் அரசியல் தலைமையிடமாக திருச்சிராப்பள்ளியை தமிழ்நாட்டின் தலைநகரமாக மாற்ற முயற்சிகள் மேற்கொள்ளப்பட்டன. ஆனால் அவரது மறைவிற்குப் பிறகு அத்திட்டம் நிறைவேற்ற முடியவில்லை. ஆனால் இன்றளவும் திருச்சியைத் தமிழ்நாட்டின் இரண்டாவது தலைநகராக்க தகுதி உடைய நகரமாக மாற்ற கோரிக்கை மக்கள் மத்தியில் ஓர் ஏக்கமாகவே காணமுடிகிறது. மற்றும் திருச்சிராப்பள்ளி பெறுநகர மாநகராட்சி என்று தரம் உயர்த்தி கூடிய விரைவில் அறிவிக்கபட இருக்கிறது.
 
 == பெயர்க்காரணம் ==
 திருச்சி மலைக்கோட்டையில் காணப்படும் [[குகை]]யில் 'சிரா' என்னும் பெயருடைய சமணத் துறவி தங்கியிருந்து தவமிருந்ததாக அக்குகையில் உள்ள பதினோராம் நூற்றாண்டுக் கல்வெட்டு கூறுகிறது. &quot;சிரா&quot; துறவியின் பள்ளி, சிராப்பள்ளி என்றாகி அதுவே இவ்வூருக்குப் பெயராகி உள்ளது என்றும் கருதப்படுகிறது.&lt;ref name=&quot;இந்து&quot;&gt;{{cite web | url=http://tamil.thehindu.com/society/kids/%E0%AE%8E%E0%AE%AA%E0%AF%8D%E0%AE%AA%E0%AE%9F%E0%AE%BF%E0%AE%AF%E0%AE%BF%E0%AE%B0%E0%AF%81%E0%AE%A8%E0%AF%8D%E0%AE%A4%E0%AE%A4%E0%AF%81-%E0%AE%85%E0%AE%A8%E0%AF%8D%E0%AE%A4%E0%AE%95%E0%AF%8D-%E0%AE%95%E0%AE%BE%E0%AE%B2%E0%AE%AA%E0%AF%8D-%E0%AE%AA%E0%AE%B3%E0%AF%8D%E0%AE%B3%E0%AE%BF%E0%AE%95%E0%AF%8D%E0%AE%95%E0%AF%82%E0%AE%9F%E0%AE%AE%E0%AF%8D/article9721322.ece | title=எப்படியிருந்தது அந்தக் காலப் பள்ளிக்கூடம்? | publisher=தி இந்து | work=கட்டுரை | date=7 சூன் 2017 | accessdate=7 சூன் 2017 | author=ஆதி}}&lt;/ref&gt; 16 ஆம் நூற்றாண்டைச் சேர்ந்த [[கல்வெட்டு]] ஒன்றில் திருச்சிராப்பள்ளி, ''திரு-சிலா-பள்ளி'' (பொருள்: &quot;புனித-பாறை-ஊர்&quot;) எனக் குறிப்பிடப்பட்டுள்ளதால் அதிலிருந்து இப்பெயர் வந்திருக்கலாம் என்ற கருத்தும் உள்ளது.&lt;ref name=&quot;hemingwayp2&quot; /&gt;&lt;ref name=&quot;hobsonjobson&quot; /&gt; வேறு சில அறிஞர்கள் &quot;திரு-சின்ன-பள்ளி&quot; (புனித-சிறிய-ஊர்) என்பதிலிருந்தும் உருவாகியிருக்கலாம் என்ற கருத்தினைக் கொண்டுள்ளனர்.&lt;ref name=&quot;hemingwayp2&quot; /&gt;&lt;ref&gt;{{cite book|last=Hemingway|first=Frederick Ricketts|title=Madras District Gazetteers: Trichinopoly|url=https://books.google.com/books?id=AtBrlwEACAAJ|year=1907|volume=1|publisher=Government Press|access-date=15 July 2016|archive-date=2 May 2014|archive-url=https://web.archive.org/web/20140502032552/http://books.google.com/books?id=AtBrlwEACAAJ|url-status=live}}&lt;/ref&gt;
 
 == வரலாறு ==
 [[படிமம்:Tiruchirappalli(Trichinopoly)-during-CarnaticWars1751.jpg|thumb|left|120px|[[கர்நாடகப் போர்கள்|கர்நாடகப்போரின்]] போது திருச்சி, 1751]]
 [[படிமம்:Tiruchirapalli1955 hicontrast.png|thumb|left|120px|திருச்சி,1955]]
 '''''முதன்மைக் கட்டுரை''''': [[திருச்சிராப்பள்ளியின் வரலாறு]]
 
 திருச்சிராப்பள்ளி, தமிழ்நாட்டில் மக்கள் வாழ்ந்த மிகப் பழமையான நகரங்களில் ஒன்று. இந்நகரத்தின் வரலாறு [[பொது ஊழி|பொ.ஊ.மு.]] இரண்டாயிரம் ஆண்டு காலத்துக்கு முந்தையது.&lt;ref name=&quot;thaninayagamp324&quot;&gt;[[#Thani Nayagam|Thani Nayagam]], p 324&lt;/ref&gt; [[முற்காலச் சோழர்கள்|முற்கால சோழர்களின்]] தலைநகராக, பொ.ஊ.மு. 3-ஆம் நூற்றாண்டு முதல் [[பொது ஊழி|பொ.ஊ.]] 3-ஆம் நூற்றாண்டு வரை விளங்கிய '''உறையூர்'''&lt;ref name=&quot;sastrip22&quot;&gt;[[#Sastri|Sastri]], p 22&lt;/ref&gt; தற்போதைய திருச்சிராப்பள்ளியின் சுற்றுப்புறத்தில் அமைந்துள்ள பகுதியாக அடையாளம் காணப்பட்டுள்ளது.&lt;ref name=&quot;sastrip19&quot;&gt;[[#Sastri|Sastri]], p 19&lt;/ref&gt;&lt;ref name=&quot;beckp40&quot;&gt;[[#Beck|Beck]], p 40&lt;/ref&gt;&lt;ref&gt;[http://www.thehindu.com/todays-paper/tp-features/tp-editorialfeatures/article2231702.ece உறையூர் சோழர் தலைநகரம் - தி இந்து]&lt;/ref&gt;&lt;ref&gt;{{Cite web |url=http://www.hindu.com/fr/2004/11/05/stories/2004110503201000.htm |title=காப்பகப்படுத்தப்பட்ட நகல் |access-date=2012-08-29 |archive-date=2005-02-08 |archive-url=https://web.archive.org/web/20050208020318/http://www.hindu.com/fr/2004/11/05/stories/2004110503201000.htm |url-status=dead }}&lt;/ref&gt; [[கரிகால் சோழன்]] கட்டிய உலகின் பழைய அணையான [[கல்லணை]] &lt;ref&gt;{{Cite web |url=http://www.thehindu.com/life-and-style/kids/article2408778.ece |title=காப்பகப்படுத்தப்பட்ட நகல் |access-date=2012-08-29 |archive-date=2012-07-17 |archive-url=https://web.archive.org/web/20120717220026/http://www.thehindu.com/life-and-style/kids/article2408778.ece |url-status=dead }}&lt;/ref&gt;&lt;ref&gt;[http://www.britannica.com/EBchecked/topic/100477/Kaveri-River பிரித்தானிகா கல்லணை கட்டப்பட்டது 2 ஆம் நூற்றாண்டு]&lt;/ref&gt; உறையூரில் (உறையூருக்கு '''கோழியூர்''' என்ற பெயரும் உண்டு) இருந்து 20 கி.மீ. தொலைவில் உள்ளது.
 
 பொ.ஊ. 5-ஆம் நூற்றாண்டில் இந்நகரம் [[பல்லவர்கள்|பல்லவர்களின்]] ஆட்சியின் கீழ் வந்தது. 6-ஆம் நூற்றாண்டில் தென்இந்தியாவை ஆண்ட [[முதலாம் மகேந்திரவர்ம பல்லவன்|முதலாம் மகேந்திரவர்மன்]] மலைக்கோட்டையில் பல குடைவரைக் கோவில்களைக் கட்டினார்.&lt;ref name=&quot;kuppuramp105&quot;&gt;[[#Kuppuram|Kuppuram]], p 105&lt;/ref&gt;&lt;ref name=&quot;sastrip105&quot;&gt;[[#Sastri|Sastri]], p 105&lt;/ref&gt;&lt;ref name=&quot;beckp42&quot;&gt;[[#Beck|Beck]], p 42&lt;/ref&gt;&lt;ref name=&quot;beckp92&quot;&gt;[[#Beck|Beck]], p 92&lt;/ref&gt; பல்லவர்களின் வீழ்ச்சிக்குப் பிறகு [[பிற்கால சோழர்கள்]] இந்நகரை 13-ஆம் நூற்றாண்டு வரை ஆட்சி புரிந்தார்கள்.&lt;ref name=&quot;sastrip438&quot;&gt;[[#Sastri|Sastri]], p 438&lt;/ref&gt; சோழர்களின் வீழ்ச்சிக்குப் பின், இந்நகரம் [[பாண்டியர்]]களின் ஆளுகைக்கு கீழ் வந்தது. 1216 முதல் 1311 வரை அவர்கள் ஆண்டார்கள். 1311ல் [[மாலிக் காபூர்]] பாண்டியர்களைத் தோற்கடித்து இந்நகரைக் கைப்பற்றினார்.&lt;ref name=&quot;aiyangarp99&quot;&gt;[[#Aiyangar|Aiyangar]], p 99&lt;/ref&gt;&lt;ref name=&quot;lal&quot;&gt;[[#Lal|Lal]], pp 251-252&lt;/ref&gt; மாலிக் காபூரின் [[தில்லி சுல்தானகம்|டில்லி சுல்தானின்]] படை பல விலைமதிக்க முடியாத பொருட்களைக் கைப்பற்றினார்கள். இவர்கள் அரங்கநாதன் கோவிலைக் களங்கப்படுத்தினதால், அக்கோயில் 60 ஆண்டுகளுக்கு மூடப்பட்டு இருந்தது.&lt;ref&gt;http://www.templenet.com/Tamilnadu/df001.html&lt;/ref&gt; வீதி உலாவிற்கு எடுத்துச்செல்லப்படும் அரங்கன் திருவுருவம் மாலிக்காபூரின் படையெடுப்பின் காரணமாகத் திருமலையில் பாதுகாக்கப்பட்டது.&lt;ref&gt;http://news.oneindia.in/2006/07/16/srirangam-temple-to-honour-tirupathi-temple-with-vasthirams-1153040574.html{{Dead link|date=அக்டோபர் 2022 |bot=InternetArchiveBot }}&lt;/ref&gt; முசுலிம்களின் ஆட்சிக்குப்பிறகு இந்நகரம் [[விஜயநகரப் பேரரசு|விஜயநகர பேரரசின்]] கீழ் வந்தது. பின்பு அவரின் இப்பகுதியின் ஆளுநர் [[மதுரை நாயக்கர்கள்|மதுரை நாயக்கர்களின்]] கீழ் 1736 வரை இருந்தது.&lt;ref name=&quot;madurainayaksp234&quot;&gt;[[#Sathianathaier|Sathianathaier]], p 234&lt;/ref&gt;.
 
 == திருச்சிராப்பள்ளியிலுள்ள அரண்மனை ==
 மதுரை நாயக்கர்களின் வழியில் வந்த [[இராணி மங்கம்மாள்]] திருச்சி மலைக்கோட்டை பகுதியில் கட்டிய அரண்மனை அரசு அருங்காட்சியகமாகச் செயல்பட்டு வருகின்றது.
 
 == மக்கள் வகைப்பாடு ==
 {{bar box|title=மதவாரியான கணக்கீடு|titlebar=#ddd|left1=மதம்|right1=சதவீதம்(%)|float=left|bars={{bar percent|[[இந்து|இந்துக்கள்]]|Orange|74.07}}{{bar percent|[[முஸ்லிம்|முஸ்லிம்கள்]]|Green|14.72}}{{bar percent|[[கிறிஸ்தவம்|கிறிஸ்தவர்கள்]]|purple|10.89}}{{bar percent|[[சைனம்|சைனர்கள்]]|Blue|0.07}}{{bar percent|[[சீக்கியம்|சீக்கியர்கள்]]|yellow|0.02}}{{bar percent|[[பௌத்தம்|பௌத்தர்கள்]]|Gold|0.01}}{{bar percent|மற்றவை|grey|0.22}}{{bar percent|[[சமயமின்மை|சமயமில்லாதவர்கள்]]|violet|0.01}}}}
 
 {{Historical populations
 | align=left
 |state=collapsed
 |1871 | 76530
 |1881 | 84449
 |1891 | 90609
 |1901 | 104721
 |1911 | 123512
 |1921 | 120422
 |1931 | 142843
 |1941 | 159566
 |1951 | 218921
 |1961 | 249862
 |1971 | 307400
 |1981 | 362045
 |1991 | 387223
 |2001 | 752066
 |2011 | 847387
 |footnote = Sources:
 * 1871&amp;nbsp;– 1901:{{Sfn|Burn|Cotton|1908|p=43}}
 * 1891&amp;nbsp;– 1941:{{Sfn|Census of India|1941}}
 * 1941&amp;nbsp;– 1961:{{Sfn|Report|1966|p=226}}
 * 1961–1991:{{Sfn|Rajendran|Arumugam|Chandrasekaran|2002|p=3}}
 * 1991–2001:{{Sfn|Malayala Manorama|2007|p=707}}
 * 2011:&lt;ref name=CMA&gt;{{cite web|title=Proceedings of the Commissioner of Municipal Administration |url=http://cma.tn.gov.in/cma/en-in/Downloads/cma1.pdf |publisher=Commissionerate of Municipal Administration |accessdate=1 October 2015 |page=4 |format=PDF |date=22 July 2014 |url-status=live |archiveurl=https://web.archive.org/web/20151002195006/http://cma.tn.gov.in/cma/en-in/Downloads/cma1.pdf |archivedate=2 October 2015 }}&lt;/ref&gt;
 }}
 
 இந்திய 2011 மக்கள் தொகைக் கணக்கெடுப்பின்படி, இந்நகரின் மக்கள்தொகை 8,47,387 ஆகும்.&lt;ref name=&quot;census&quot;&gt;{{cite web |  accessdate = ஜனவரி 30, 2007 |  url = http://www.censusindia.net/results/town.php?stad=A&amp;state5=999 |  title = 2011-ஆம் ஆண்டிற்கான இந்திய மக்கள்தொகைக் கணக்கெடுப்பு அறிக்கை |  archive-date = 2004-06-16 |  archive-url = https://web.archive.org/web/20040616075334/http://www.censusindia.net/results/town.php?stad=A&amp;state5=999 |  url-status = unfit }}&lt;/ref&gt; இவர்களில் 50% ஆண்கள், 50% பெண்கள் ஆவார்கள். திருச்சிராப்பள்ளி மக்களின் சராசரி கல்வியறிவு 91.38% ஆகும், இதில் ஆண்களின் கல்வியறிவு 94.85% , பெண்களின் கல்வியறிவு 88.01% ஆகும். இது இந்தியத் தேசிய சராசரிக் கல்வியறிவான 59.5% விடக் கூடியதாகும். திருச்சிராப்பள்ளி மக்கள் தொகையில் 10% ஆறு வயதுக்குட்பட்டோர் ஆவார்கள்.
 
 தமிழ்நாட்டில் நான்காவது மிகப்பெரும் மாநகரப்பகுதியாக விளங்கும் திருச்சிராப்பள்ளி, இந்தியளவில் 51வது இடத்தில் உள்ளது. 2011இல் நடந்த மக்கள்தொகை கணக்கெடுப்பின் துவக்கநிலை மதிப்பீடுகளின்படி மாநகரப்பகுதியின் மக்கள்தொகை 8,47,387 ஆகவும்,&lt;ref name=&quot;metro_2011provisionaltotals&quot;&gt;{{cite web|title=Table 2: Cities having population 1 lakh and above|work=Provisional Population Totals|publisher=Government of India|url=http://www.censusindia.gov.in/2011-prov-results/paper2/data_files/India2/Table_2_PR_Cities_1Lakh_and_Above.pdf|accessdate=2011-10-19|format=PDF}}&lt;/ref&gt; கூட்டுநகரப்பகுதியின் மக்கள்தொகை 10,22,518 ஆகவும் உள்ளது.&lt;ref name=&quot;ua_2011provisionaltotals&quot;&gt;{{cite web|title=Table 3: Urban agglomerations having population 1 Lakh and above|url=http://www.censusindia.gov.in/2011-prov-results/paper2/data_files/India2/Table_3_PR_UA_Citiees_1Lakh_and_Above.pdf|work=Provisional Population Totals|publisher=Government of India|accessdate=2011-10-19|format=PDF}}&lt;/ref&gt; திருச்சிராப்பள்ளியில் 1,62,000 மக்கள் 286 குடிசைப்பகுதிகளில் வாழ்கின்றனர்.&lt;ref name=&quot;wateraid&quot;&gt;{{cite book|title=Tiruchirappalli shows the way|publisher=Wateraid India|year=2008|url=http://www.wateraid.org/documents/plugin_documents/tiruchirappalli_shows_the_way.pdf|page=4|accessdate=2011-05-13|format=PDF}}&lt;/ref&gt;
 
 மக்கள்தொகையில் [[இந்து]]க்கள் பெரும்பான்மையினராக இருப்பினும் கணிசமான அளவில் [[கிறித்தவம்|கிறித்தவர்களும்]]&lt;ref name=&quot;1971censusp88&quot;&gt;{{cite book|title=Census of India, 1971: Tamil Nadu|first=K.|last=Chockalingam|page=88|publisher=Government of India|year=1979}}&lt;/ref&gt; [[முசுலிம்]]களும் வாழ்கின்றனர்.&lt;ref name=&quot;thehindu_20081210&quot;&gt;{{cite news|title=Bakrid celebrated in Trichy with gaiety|url=http://www.hindu.com/2008/12/10/stories/2008121059380300.htm|work=The Hindu|location=India|date=10 December 2008|accessdate=2011-05-11|archivedate=2008-12-14|archiveurl=https://web.archive.org/web/20081214152149/http://www.hindu.com/2008/12/10/stories/2008121059380300.htm|url-status=dead}}&lt;/ref&gt; குறைந்த எண்ணிக்கையில் [[சீக்கியர்]]களும்&lt;ref name=&quot;thehindu_20101229&quot;&gt;{{cite news|title=Lt. Governor felicitated|work=The Hindu|location=India|date=29 December 2010|url=http://www.hindu.com/2010/12/29/stories/2010122964040300.htm|accessdate=2011-05-11|archivedate=2011-01-03|archiveurl=https://web.archive.org/web/20110103033615/http://www.hindu.com/2010/12/29/stories/2010122964040300.htm|url-status=dead}}&lt;/ref&gt; [[சமணம்|சமணர்களும்]]&lt;ref name=&quot;thehindu_20110407&quot;&gt;{{cite news|title=Jain Sangh celebrates Mahaveer Jayanthi|work=The Hindu|location=India|date=17 April 2011|url=http://www.hindu.com/2011/04/17/stories/2011041750960200.htm|accessdate=2011-05-11|archivedate=2011-04-23|archiveurl=https://web.archive.org/web/20110423023500/http://www.hindu.com/2011/04/17/stories/2011041750960200.htm|url-status=dead}}&lt;/ref&gt;  இங்குள்ளனர். மிகப் பரவலாகப் பேசப்படும் மொழியாகத் [[தமிழ்]] விளங்கினாலும்&lt;ref name=&quot;tourisminindiap606&quot;&gt;[[#Sen|Sen]], p 606&lt;/ref&gt; கணிசமான மக்கள் [[தெலுங்கு மொழி|தெலுங்கு]],&lt;ref name=&quot;teluguassocp116&quot;&gt;[[#Ramappa and Sudershan|Ramappa and Sudershan]], p 116&lt;/ref&gt; [[சௌராட்டிர மொழி]]&lt;ref name=&quot;unwritten&quot;&gt;{{cite news|title=The Unwritten History of the Saurashtrians of South India|first=T. A.|last=Ramesh|url=http://www.boloji.com/history/039.htm|date=15 January 2006|publisher=boloji.com|access-date=2 செப்டம்பர் 2012|archivedate=14 ஜனவரி 2010|archiveurl=https://web.archive.org/web/20100114184221/http://boloji.com/history/039.htm|url-status=dead}}&lt;/ref&gt; மற்றும் [[கன்னட மொழி]]&lt;ref name=&quot;karnatakaiii&quot;&gt;[[#Muthanna|Muthanna]], Preface, p iii&lt;/ref&gt; பேசுகின்றனர்.  சௌராட்டிர மொழியை 16வது நூற்றாண்டில் [[குசராத்]]திலிருந்து குடிபெயர்ந்து வாழும் பட்டு நூல்காரர்கள் தாய்மொழியாகக் கொண்டுள்ளனர்.&lt;ref name=&quot;thurstonp123&quot;&gt;{{cite book|title=Provincial Geographies of India Vol 4: The Madras Presidency with Mysore, Coorg and Associated States|last=Thurston|first=Edgar|publisher=Cambridge University|year=1913|page=123}}&lt;/ref&gt; மேலும் இங்கு குறிப்பிடத்தக்க அளவில் [[இலங்கைத் தமிழர்|புலம் பெயர்ந்த இலங்கைத் தமிழர்]] நகரத்தின் சுற்றுப்புறங்களில் அமைக்கப்பட்டுள்ள ஏதிலிகள் முகாம்களில் வாழ்கின்றனர்.&lt;ref name=&quot;thehindu_20091103&quot;&gt;{{cite news|title=Sri Lankan Tamil refugees wish for Indian citizenship|url=http://www.thehindu.com/news/states/tamil-nadu/article42734.ece|work=The Hindu|location=India|date=3 November 2009|accessdate=2011-05-11|archivedate=2012-03-23|archiveurl=https://web.archive.org/web/20120323011206/http://www.thehindu.com/news/states/tamil-nadu/article42734.ece|url-status=dead}}&lt;/ref&gt; இங்குள்ள தென்னக இரயில்வேயின் மண்டல தலைமையகத்தையொட்டி கணிசமான [[ஆங்கிலோ இந்தியர்]]கள் வாழ்கின்றனர்.&lt;ref name=&quot;anglochroniclesp205&quot;&gt;{{cite book|title=The way we were: Anglo-Indian chronicles|page=205|first1=Glenn|last1=Deefholts|first2=Quentine|last2=Acharya|publisher=Calcutta Tiljallah Relief Inc|year=2006|id={{ISBN|0-9754639-3-4}}, {{ISBN|978-0-9754639-3-2}}}}&lt;/ref&gt;
 
 == புவியியலும் வானிலையும் ==
 {{climate chart
 |திருச்சிராப்பள்ளி
 |20.3|30.1|14.3
 |20.9|32.6|5.4
 |23|35.1|9.5
 |25.8|36.9|50.5
 |26.4|37.4|65.2
 |26.5|36.7|34.9
 |25.9|35.7|60.6
 |25.5|35.4|85.5
 |24.7|34.5|146.6
 |23.9|32.2|191.5
 |22.7|30.1|131.8
 |21.2|29.2|84.4
 |source=இந்திய வானிலையியல் துறை&lt;ref name=&quot;indianmeteo&quot;&gt;{{cite web|title=Climatology of Tiruchirappalli|url=http://www.imd.gov.in/section/climate/tiruchirapalli2.htm|publisher=இந்திய வானிலையியல் துறை|accessdate=2011-05-12}}&lt;/ref&gt; &lt;br /&gt;வரைபட எளிமைக்காக, வரைபடத்திலுள்ள பொழிவு எண்கள் &lt;br /&gt;  1:10 அளவில் சுருக்கப்பட்டுள்ளன.
 |float=right}}
 [[படிமம்:Rockfort View of Trichy.jpg|thumb|left|250px|மலைக்கோட்டையின் மேலிருந்து திருச்சி மாநகரின் வான்வழித் தோற்றம்]]
 திருச்சிராப்பள்ளி {{Coord|10.8050|N|78.6856|E|}} என்ற புவியியல் ஆள்கூறுகளில் அமைந்துள்ளது.&lt;ref name=&quot;fallinggrain&quot;&gt;{{cite web|title=Tiruchirappalli|url=http://www.fallingrain.com/world/IN/25/Tiruchchirappalli.html|publisher=Falling Grain genomics Inc.|accessdate=2011-05-12}}&lt;/ref&gt; நகரத்தின் சராசரி உயரம் {{convert|88|m|ft}} ஆகும்.&lt;ref name=&quot;weatherbase&quot;&gt;{{cite web|title=Weather data for Tiruchchirappalli, India|url=http://www.weatherbase.com/weather/weather.php3?s=44334&amp;refer=|publisher=Canty and Associates LLC|accessdate=2011-05-12}}&lt;/ref&gt; இது தமிழ்நாட்டின் புவியியல் மையத்திற்கு அண்மையில் அமைந்துள்ளது. இதன் தரைப்பகுதி பெரும்பாலும் தட்டையாகச் சிற்சில இடங்களில் அங்கொன்றும் இங்கொன்றுமாகக் குன்றுகளுடன்  காணப்படுகிறது; இந்தக் குன்றுகளிலேயே உயரமான குன்றாக  '''மலைக்கோட்டை''' விளங்குகிறது.&lt;ref name=&quot;imperialgazetteer1908v24p26&quot;&gt;[[#Imperial Gazetteer of India|Imperial Gazetteer of India]], Vol 24, p 26&lt;/ref&gt;&lt;ref name=&quot;rajendranp3&quot;&gt;{{Sfn|Rajendran|Arumugam|Chandrasekaran|2002|p=3}}&lt;/ref&gt; வடக்கில் [[சேர்வராயன் மலை]]க்கும் தெற்கு-தென்மேற்கில் [[பழனி மலை]]க்கும் இடைப்பட்ட {{Convert|146.7|km2}} பரப்பளவில் இந்த நகரம் அமைந்துள்ளது.&lt;ref name=&quot;abramp489&quot;&gt;[[#Abram|Abram]], p 489&lt;/ref&gt; திருச்சியின் மேற்கே {{Convert|16|km}} தொலைவில் [[காவிரி கழிமுகம்|காவிரியின் கழிமுகம்]] தொடங்குகிறது.&lt;ref name=&quot;pradeepsharmap317&quot;&gt;[[#Sharma|Sharma]], p 117&lt;/ref&gt; இப்பகுதியில் [[காவேரி ஆறு]] இரண்டாகப் பிரிந்து  [[திருவரங்கம்|திருவரங்கத் தீவு]] உண்டாகி உள்ளது.&lt;ref name=&quot;moorep61&quot;&gt;[[#Moore|Moore]], p 61&lt;/ref&gt;
 
 காவேரி ஆற்றையொட்டிய  பகுதிகளில் அமைந்துள்ள வயல்பகுதிகளில் காவேரி ஆறும் அதன் கிளையாறான கொள்ளிடம் ஆறும் செழிப்பான வண்டல் மண்ணைக் கொண்டு சேர்த்துள்ளன.&lt;ref name=&quot;imperialgazetteer1908v24p32&quot;&gt;[[#Imperial Gazetteer of India|Imperial Gazetteer of India]], Vol 24, p 32&lt;/ref&gt;  தெற்கில், செழிப்பு குறைந்த கருமண் நிலமாக உள்ளன.&lt;ref name=&quot;imperialgazetteer1908v24p32&quot; /&gt; வண்டல் வளமிகு நன்செய் நிலங்களில் ''ராகி''யும் ''சோளமும்'' பயிரிடப்படுகின்றன.&lt;ref name=&quot;imperialgazetteer1908v24p33&quot;&gt;[[#Imperial Gazetteer of India|Imperial Gazetteer of India]], Vol 24, p 33&lt;/ref&gt; வட-கிழக்கில் ''திருச்சிராப்பள்ளி வகை'' என்றழைக்கப்படும் [[கிரீத்தேசியக் காலம்|கிரீத்தேசிய]] பாறைகள் காணப்படுகின்றன.&lt;ref name=&quot;geologicalsurveyp104&quot;&gt;[[#Geological Survey of India|Geological Survey of India]], p 104&lt;/ref&gt; தென்-கிழக்குப் பகுதியில் மெல்லிய லாடரைட்டு பாறைகளின் கீழாகக் கிரானைட்டுக் கற்கள் கிடைக்கின்றன.&lt;ref name=&quot;imperialgazetteer1908v24p26&quot; /&gt;
 
 நகரத்தின் வட பகுதியில் தொழிற்பேட்டைகளும் வசிப்பிடங்களும் நெருக்கமாக அமைந்துள்ளன.&lt;ref name=&quot;rajendranp3&quot;/&gt; நகரத்தின் தென்பகுதியிலும் குடியிருப்புப் பகுதிகள் நிறைந்துள்ளன.&lt;ref name=&quot;rajendranp3&quot; /&gt; நகரைச் சுற்றிலும் வயல்வெளிகள் சூழ்ந்துள்ளன.&lt;ref name=&quot;rajendranp3&quot; /&gt; கோட்டைக்குள் அடங்கியுள்ள பழைய நகரம் திட்டமிடப்படாததாகவும் நெரிசல்மிக்கதாகவும் விளங்குகிறது. புதிய நகர்ப்புறப்பகுதிகள் திட்டமிடப்பட்டுக் கட்டமைக்கப்பட்டுள்ளன.&lt;ref name=&quot;census1961p215&quot;&gt;[[#Superintendent, Census Operations|Superintendent, Census Operations]], p 215&lt;/ref&gt; ஸ்ரீரங்கத்தில் அமைந்துள்ள பல வீடுகள் தொன்மையான இந்து சமய கோவில் வடிவமைப்புக்களுக்கான சிற்ப சாத்திரங்களுக்கேற்ப வடிவமைக்கப்பட்டுள்ளன.&lt;ref name=&quot;Ayyarp453&quot;&gt;[[#Ayyar|Ayyar]], p 453&lt;/ref&gt; தமிழ்நாடு  நகர மற்றும் ஊரகத் திட்டச்சட்டம் 1974 இக்கு இணங்க ஏப்ரல் 5, 1974 இல் திருச்சிராப்பள்ளி நகர திட்ட ஆணையம் உருவாக்கப்பட்டது.&lt;ref name=&quot;tiruchi_lpa&quot;&gt;{{cite web|title=Tiruchirappalli Local Planning Authority&amp;nbsp;– An organisation|url=http://tiruchirappalli.nic.in/tlpa/organisation.htm|publisher=Tiruchirappalli Local Planning Authority|accessdate=2011-10-19}}&lt;/ref&gt;  காவேரி ஆற்றிலிருந்து நகரத்திற்கு 1,470 நிலத்தடி நீரேற்றிகள், 60 வழங்கல் நீர்த்தொட்டிகள் மூலமாகக் குடிநீர் வழங்கப்படுகிறது.&lt;ref name=&quot;wateraid&quot; /&gt;
 
 ஆண்டின் எட்டு மாதங்களுக்காவது திருச்சி வெப்பமிகுந்து ஈரப்பதம் குறைந்து காணப்படுகின்றது.&lt;ref name=&quot;annesley&quot;&gt;[[#Annesley|Annesley]], pp 62-64&lt;/ref&gt; மார்ச்சு முதல் சூலை வரை மிகவும் வெப்பமான வானநிலை நிலவுகிறது.&lt;ref name=&quot;annesley&quot; /&gt;&lt;ref name=&quot;moorep88&quot;&gt;[[#Moore|Moore]], p 88&lt;/ref&gt; ஆகத்து முதல் அக்டோபர் வரை பெருங்காற்றுடன் கூடிய இடிமழையுடன் மிதமான வானிலை நிலவுகிறது. நவம்பர் முதல் பெப்ரவரி வரை குளிர்காலமாக விளங்குகிறது.&lt;ref name=&quot;annesley&quot; /&gt; பனிமூட்டமும் பனித்துளிகளும் அரிதானவை; அவை குளிர் மாதங்களில் ஏற்படலாம்.&lt;ref name=&quot;annesley&quot; /&gt;
 
 {{Weather box
 | location = திருச்சிராப்பள்ளி
 | metric first = Yes
 | single line = Yes
 | Jan record high C = 35.6
 | Feb record high C = 40.0
 | Mar record high C = 42.2
 | Apr record high C = 42.8
 | May record high C = 43.3
 | Jun record high C = 43.9
 | Jul record high C = 41.1
 | Aug record high C = 40.6
 | Sep record high C = 40.6
 | Oct record high C = 38.9
 | Nov record high C = 36.7
 | Dec record high C = 35.6
 |year record high C = 43.9
 | Jan high C = 30.3
 | Feb high C = 32.8
 | Mar high C = 35.7
 | Apr high C = 37.5
 | May high C = 38.2
 | Jun high C = 37.1
 | Jul high C = 36.3
 | Aug high C = 35.8
 | Sep high C = 35.0
 | Oct high C = 32.8
 | Nov high C = 30.4
 | Dec high C = 29.6
 |year high C = 34.3
 | Jan low C = 20.6
 | Feb low C = 21.5
 | Mar low C = 23.5
 | Apr low C = 26.1
 | May low C = 26.8
 | Jun low C = 26.6
 | Jul low C = 26.1
 | Aug low C = 25.7
 | Sep low C = 24.9
 | Oct low C = 24.2
 | Nov low C = 22.8
 | Dec low C = 21.2
 |year low C = 24.2
 | Jan record low C = 14.4
 | Feb record low C = 13.9
 | Mar record low C = 15.6
 | Apr record low C = 18.3
 | May record low C = 19.4
 | Jun record low C = 18.0
 | Jul record low C = 20.1
 | Aug record low C = 20.6
 | Sep record low C = 20.6
 | Oct record low C = 18.9
 | Nov record low C = 16.7
 | Dec record low C = 14.4
 |year record low C = 13.9
 | rain colour = green
 | Jan rain mm = 13.3
 | Feb rain mm = 3.6
 | Mar rain mm = 5.3
 | Apr rain mm = 29.6
 | May rain mm = 67.0
 | Jun rain mm = 38.3
 | Jul rain mm = 60.5
 | Aug rain mm = 69.9
 | Sep rain mm = 153.4
 | Oct rain mm = 153.9
 | Nov rain mm = 168.0
 | Dec rain mm = 81.4
 |year rain mm = 844.2
 | Jan rain days = 0.9
 | Feb rain days = 0.4
 | Mar rain days = 0.6
 | Apr rain days = 1.7
 | May rain days = 3.8
 | Jun rain days = 2.6
 | Jul rain days = 2.7
 | Aug rain days = 4.1
 | Sep rain days = 7.1
 | Oct rain days = 9.2
 | Nov rain days = 8.1
 | Dec rain days = 5.1
 |year rain days = 46.3
 |time day = 17:30 [[Indian Standard Time|IST]]
 | Jan humidity = 54
 | Feb humidity = 43
 | Mar humidity = 37
 | Apr humidity = 41
 | May humidity = 42
 | Jun humidity = 44
 | Jul humidity = 45
 | Aug humidity = 46
 | Sep humidity = 52
 | Oct humidity = 64
 | Nov humidity = 70
 | Dec humidity = 65
 |year humidity = 50
 |source 1 = [[இந்திய வானிலை ஆய்வுத்துறை]]&lt;ref name=IMDextremes&gt;
 {{cite web
 | archive-url = https://web.archive.org/web/20200205042509/http://imdpune.gov.in/library/public/EXTREMES%20OF%20TEMPERATURE%20and%20RAINFALL%20upto%202012.pdf
 | archive-date = 5 February 2020
 | url = https://imdpune.gov.in/library/public/EXTREMES%20OF%20TEMPERATURE%20and%20RAINFALL%20upto%202012.pdf
 | title = Extremes of Temperature &amp; Rainfall for Indian Stations (Up to 2012)
 | publisher = India Meteorological Department
 | date = December 2016
 | page = M204
 | access-date = 30 March 2020}}&lt;/ref&gt;&lt;ref name=IMDnormals&gt;
 {{cite web
 | archive-url = https://web.archive.org/web/20200205040301/http://imdpune.gov.in/library/public/1981-2010%20CLIM%20NORMALS%20%28STATWISE%29.pdf
 | archive-date = 5 February 2020
 | url = https://imdpune.gov.in/library/public/1981-2010%20CLIM%20NORMALS%20%28STATWISE%29.pdf
 | title = Station: Tiruchirapalli (A) Climatological Table 1981–2010
 | work = Climatological Normals 1981–2010
 | publisher = India Meteorological Department
 | date = January 2015
 | pages = 745–746
 | access-date = 30 March 2020}}&lt;/ref&gt;
 |date=May 2012}}
 {{clear}}
 
 == தலைநகராக்க முயற்சி ==
 {{main|திருச்சியைத் தலைநகராக மாற்றும் திட்டம்}}
 {{wide image|Trichy pano7.jpg|1800px|திருச்சிராப்பள்ளியின் காவிரி நதியையும், ஸ்ரீரங்கம் தீவையும் காட்டும் அகலப் பரப்பு புகைப்படம். (இப்புகைப்படத்தை முழுமையாக பார்க்க, இப்படத்தை வலமிருந்து, இடப்புறமாக இழுக்கவும்)}}
 [[File:Aerial photograph of Srirangam Island between Kaveri and Kollidam rivers.JPG|thumb|right|250px|ஸ்ரீரங்கம் தீவின் வான்வழி புகைப்படம்]]
 
 [[எம். ஜி. ஆர்|எம்.ஜி.ஆர்]] [[தமிழ்நாடு|தமிழ்நாட்டின்]] முதல்வராக இருந்த போது அதன் தலைநகரமாகத் திருச்சியை மாற்றும் திட்டத்தினை வழிவகுத்தார். எதிர்க்கட்சிகளின் எதிர்ப்பினாலும் அரசியல் சூழ்நிலையாலும் இத்திட்டம் நிறைவேற்றப்படவில்லை.
 
 தமிழகத்தின் ஒரு முனையில் இருக்கும் [[சென்னை]]யின் தலைமைச் செயலகத்துக்குத்  தெற்குபகுதியில் உள்ள மக்கள் வருவது சிரமமாக உள்ளது. இதனால் தமிழ்நாட்டில் உள்ள அனைத்து மக்களும் [[தலைமைச் செயலகம் (தமிழ்நாடு)|தலைமைச் செயலகத்துக்கு]] எளிதாக வரத் தமிழகத்தின் நடுமையத்தில் இருக்கக்கூடிய திருச்சியைத் தலைநகரமாக மாற்றவேண்டும் என்று எம்.ஜி.ஆர் கருதினார். இதற்காக 1983 இல் திருச்சியைத் தலைநகரமாக்கும் திட்டத்தை அறிவித்தார்.&lt;ref&gt;[http://www.dinamalar.com/News_Detail.asp?Id=61525 தமிழகத்தின் புதிய தலைநகரம் &quot;திருச்சி': எம்.ஜி.ஆர்., கனவை நிறைவேற்றுவரா ஜெ]&lt;/ref&gt;
 
 == மாநகராட்சி நிர்வாகம் ==
 [[படிமம்:trichycorporation.jpg|right|thumb|திருச்சிராப்பள்ளி மாநகராட்சி]] 
 திருச்சிராப்பள்ளி மாநகராட்சி 100 வார்டுகளைக் கொண்டுள்ளது. ஒவ்வொரு மண்டலத்திற்கும் பகுதிகள் வீதம் [[பொன்மலை]], [[அபிஷேகபுரம்]], [[ஸ்ரீரங்கம்]], [[அரியமங்கலம்]], [[துவாக்குடி]], [[திருவெறும்பூர்]] என ஆறு மண்டலங்களாகப் பிரிக்கப்பட்டுள்ளன. இங்கு மாநகராட்சித் தலைவர் (மேயர்) தலைமையிலான மாநகர்மன்றக் கூட்டத்தில் நிறைவேற்றப்படும் தீர்மானங்களின் அடிப்படையில் பாதுகாக்கப்பட்ட குடிநீர் விநியோகம், பொது சுகாதாரம் பேணுதல், சாலைகள் பராமரிப்பு, மழைநீர் வடிகால், தெரு விளக்குகள் போன்ற இன்றியமையாத அடிப்படை வசதிகள் (மாநகராட்சிப் பணிகள்) மாநகராட்சி ஆணையரின் வழியாக, மாநகராட்சி அதிகாரிகள் மற்றும் ஊழியர்களால் செயல்படுத்தப்படுகிறது.
 
 == பொருளாதாரம் ==
 [[படிமம்:கேப்ஜெமினி.jpg|250px|thumb|[[கேப்ஜெமினி|கேப்ஜெமினி மென்பொருள் நிறுவனம்]], கரூர் ரோடு, [[திருச்சி]] ]]
 பிரித்தானிய ஆட்சிக்காலத்தில் திருச்சிராப்பள்ளி எண்ணெய் செக்குகள், தோல் பதனிடும் தொழில் மற்றும் சுருட்டு தயாரிப்பிற்குப் புகழ் பெற்றிருந்தது.&lt;ref name=&quot;imperialgazetteer1908v24p36&quot;&gt;[[#Imperial Gazetteer of India|Imperial Gazetteer of India]], Vol 24, p 36&lt;/ref&gt; உச்சத்தில் இருந்த நேரத்தில் ஆண்டுக்கு12 மில்லியன் சுருட்டுகள் தயாரிக்கப்பட்டு ஏற்றுமதியாயின.&lt;ref name=&quot;imperialgazetteer1908v24p36&quot; /&gt; பதனிடப்பட்ட தோல்கள் இங்கிருந்து ஐக்கிய இராச்சியத்திற்கு ஏற்றுமதி ஆயின.&lt;ref name=&quot;imperialgazetteer1908v24p36&quot; /&gt; திருச்சி நகரில் ஏராளமான சில்லறை மற்றும் மொத்த வணிகக் கூடங்கள் அமைந்துள்ளன; இவற்றில் ''மகாத்மா காந்தி சந்தை'' எனப்படும்  காய்கறி சந்தை மிகவும் அறியப்பட்டதாகும் &lt;ref name=&quot;thehindu_20101228&quot;&gt;{{cite news|title=Arrival of onions from Maharashtra stabilizes price|work=The Hindu|location=India|url=http://www.hindu.com/2010/12/28/stories/2010122854770800.htm|first=Syed Muthathar|last=Saqaf|date=28 December 2010|accessdate=2011-05-11|archivedate=2010-12-31|archiveurl=https://web.archive.org/web/20101231072338/http://www.hindu.com/2010/12/28/stories/2010122854770800.htm|url-status=dead}}&lt;/ref&gt;&lt;ref name=&quot;thehindu_20100601&quot;&gt;{{cite news|title=Market to be shifted|url=http://www.hindu.com/2010/06/01/stories/2010060158570100.htm|work=The Hindu|location=India|date=1 June 2010|accessdate=2011-05-11|archivedate=2010-06-06|archiveurl=https://web.archive.org/web/20100606024606/http://www.hindu.com/2010/06/01/stories/2010060158570100.htm|url-status=dead}}&lt;/ref&gt;&lt;ref name=&quot;trichytourism&quot;&gt;{{cite web|title=Trichy a land of tradition|url=http://www.tn.gov.in/trichytourism/otherother.htm|publisher=Government of Tamil Nadu|accessdate=2011-05-12}}&lt;/ref&gt; திருவரங்கத்தில் உள்ள பூக்கடைத் தெருவும்   மாம்பழச்சாலையின் மாம்பழச் சந்தையும் குறிப்பிடத்தக்கன.&lt;ref name=&quot;trichytourism&quot; /&gt;&lt;ref name=&quot;thehindu_20050504&quot;&gt;{{cite news|title=Bitter fall in mango prices, thanks to huge arrivals|url=http://www.hindu.com/2005/05/04/stories/2005050415600300.htm|work=The Hindu|location=India|date=4 May 2005|first=M.|last=Balaganessin|accessdate=2011-05-11|archivedate=2005-05-07|archiveurl=https://web.archive.org/web/20050507170630/http://www.hindu.com/2005/05/04/stories/2005050415600300.htm|url-status=dead}}&lt;/ref&gt; சுற்றுப்புற நகரான மண்ணச்சநல்லூரில் அரிசி ஆலைகளில் மெருகேற்றப்பட்ட ''[[பொன்னி அரிசி]]''  தயாராகிறது.&lt;ref name=&quot;thehindu_20081017&quot;&gt;{{cite news|title=Power holiday stifles output of rice mills|first=S.|last=Ganesan|work=The Hindu|date=17 October 2008|url=http://www.hindu.com/2008/10/17/stories/2008101754900500.htm|accessdate=2011-12-22|location=Chennai, India|archivedate=2013-10-23|archiveurl=https://web.archive.org/web/20131023060049/http://www.hindu.com/2008/10/17/stories/2008101754900500.htm|url-status=dead}}&lt;/ref&gt;
 
 இங்குள்ள நடுவண் அரசின் துப்பாக்கித் தொழிற்சாலையில் இந்திய படைத்துறைக்காக எறி குண்டுகள் உந்துகருவிகள், பல குண்டுகள் உந்து கருவிகள், துப்பாக்கிகள்  தயாரிக்கப்படுகின்றன.&lt;ref name=&quot;oft&quot;&gt;{{cite web |url=http://ofbindia.nic.in/units/index.php?unit=oft |title=Indian Ordnance Factories: Ordnance Factory Tiruchirapalli |publisher=Ordnance Factory Board, Defence Ministry, Government of India |accessdate=2011-05-12 |archive-date=2011-04-23 |archive-url=https://web.archive.org/web/20110423103517/http://ofbindia.nic.in/units/index.php?unit=oft |url-status=dead }}&lt;/ref&gt; இதே வளாகத்தில் கனத்த கலவைமாழை ஊடுருவு திட்ட (HAPP) வசதியும்&lt;ref name=&quot;happ&quot;&gt;{{cite web|title=Indian Ordnance Factories: Heavy Alloy Penetrator Project|url=http://ofbindia.nic.in/units/index.php?unit=happ|publisher=Ordnance Factory Board, Defence Ministry, Government of India|accessdate=2011-05-12|archive-date=2011-05-27|archive-url=https://web.archive.org/web/20110527001950/http://ofbindia.nic.in/units/index.php?unit=happ|url-status=dead}}&lt;/ref&gt; நடுவண் அரசின் துப்பாக்கி தொழிற்சாலைகள் வாரியத்தால் நடத்தப்படுகிறது.&lt;ref name=&quot;powdermetallurgyp50&quot;&gt;{{cite book|title=Powder metallurgy: processing for automotive, electrical/electronic and engineering industry ; [International Conference on Powder Metallurgy for Automotive and Engineering Industry ... at the Renaissance Mumbai Hotel and Convention Center during Feb. 3&amp;nbsp;– 6, 2005]|page=50|first=P.|last=Ramakrishnan|publisher=New Age International|year=2007|id={{ISBN|81-224-2030-3}}, {{ISBN|978-81-224-2030-2}}}}&lt;/ref&gt; 1980 ஆண்டுகளில் நிறுவப்பட்ட இந்த வசதியில் இந்தியாவிலேயே முதன்முறையாக ''நெகிழ்வுறு தயாரிப்பு அமைப்பு'' (FMS), பயன்படுத்தப்பட்டுள்ளது.&lt;ref name=&quot;cadp656&quot;&gt;{{cite book|title=Cad/cam/cim|first=P.|last=Radhakrishnan|publisher=New Age International|year=2000|id={{ISBN|81-224-1248-3}}, {{ISBN|978-81-224-1248-2}}|page=656}}&lt;/ref&gt;
 
 திருச்சி தமிழ்நாட்டின் பொறியியல் சாதனங்கள் தயாரிக்கும் முனையமாக விளங்குகிறது. 1928 ஆம் ஆண்டில் தொடர்வண்டி பணிப்பட்டறை [[நாகப்பட்டினம்|நாகப்பட்டினத்திலிருந்து]] திருச்சிராப்பள்ளியின் ''கோல்டன் ராக்'' (பொன்மலை) கிற்கு மாற்றப்பட்டது.  தமிழ்நாட்டின் மூன்று பட்டறைகளில் இஃது ஒன்றாகத் திகழ்கிறது.&lt;ref name=&quot;colonialismp126&quot;&gt;{{cite book|title=Children of Colonialism: Anglo-Indians in a Postcolonial World|url=https://archive.org/details/childrenofcoloni0000lion|page=[https://archive.org/details/childrenofcoloni0000lion/page/126 126]|first=Lionel|last=Caplan|publisher=Berg Publishers|year=2003|id={{ISBN|1-85973-632-7}}, {{ISBN|978-1-85973-632-6}}}}&lt;/ref&gt; இந்தப் பட்டறையிலிருந்து 2007-08 ஆண்டில் 650 வழமையான மற்றும் குறைந்த மட்ட சரக்கு வண்டிகள் தயாரிக்கப்பட்டன.&lt;ref name=&quot;goldenrock_thehindu20080411&quot;&gt;{{cite news|title=Golden Rock railway workshop rolls out stainless steel wagon prototype|url=http://www.hindu.com/2008/04/11/stories/2008041156540200.htm|first=P.|last=Rajaram|work=The Hindu|location=India|date=11 April 2008|accessdate=2011-05-11|archivedate=2008-04-14|archiveurl=https://web.archive.org/web/20080414210046/http://www.hindu.com/2008/04/11/stories/2008041156540200.htm|url-status=dead}}&lt;/ref&gt;
 
 இந்தியாவின் மிகப்பெரும் பொதுத்துறை பொறியியல் நிறுவனமாக மே 1965 இல் உயரழுத்த கொதிகலன்கள் தயாரிக்கும் [[பாரத மிகு மின் நிறுவனம்]] (BHEL), நிறுவப்பட்டது.&lt;ref name=&quot;casestudiesp71&quot;&gt;{{cite book|title=Case Studies In Marketing: The Indian Context 4Th Ed|author=Srinivasan|publisher=PHI Learning Pvt. Ltd.|id={{ISBN|81-203-3543-0}}, {{ISBN|978-81-203-3543-1}}|page=71|chapter=Bharat Heavy Electricals Limited}}&lt;/ref&gt;&lt;ref name=&quot;inflationp122&quot;&gt;{{cite book|title=Inflation accounting practices in India's corporate sector|first=Mohd Rizwan|last=Ahmad|publisher=Atlantic Publishers &amp; Distributors|year=2003|id={{ISBN|81-269-0216-7}}, {{ISBN|978-81-269-0216-3}}|page=122}}&lt;/ref&gt; இதனைத் தொடர்ந்து {{INR}}58 [[கோடி]] (US$13 மில்லியன்) செலவில் ஒட்டற்ற எஃகு ஆலையும் கொதிகலன் துணைஉதிரிகள் தொழிற்சாலையும் நிறுவப்பட்டன. இவை மூன்றும் இணைந்து {{Convert|22,927.4|m2|ft2}} பரப்பளவில் பிஎச்ஈஎல் தொழிற்சாலை வளாகமாக அறியப்படுகிறது. இங்கு நிலக்கரியைப் பயன்படுத்தி  6.2 MW மின்சாரம் தயாரிக்கப்படுகிறது.&lt;ref name=&quot;bioenergyp19&quot;&gt;{{cite book|title=Bio-energy for rural energisation: proceedings of the National Bio-Energy Convention-95 on Bio-Energy for Rural engergisation, organised by Bio-Energy Society of India, during December 4–15, 1995 at I.I.T. New Delhi|first1=R. C.|last1=Maheshwari|first2=Pradeep|last2=Chaturvedi|publisher=Concept Publishing Company|year=1997|page=19|id={{ISBN|81-7022-670-8}}, {{ISBN|978-81-7022-670-3}}}}&lt;/ref&gt; மேலும் 1961இல் ''திருச்சி எஃகு  உருட்டல் ஆலைகள்'' (''Trichy Steel Rolling Mills'') துவங்கப்பட்டது.&lt;ref name=&quot;madrasdistp539&quot;&gt;{{cite book|title=Madras District Gazetteers: Tiruchirappalli (pt. 1–2)|page=539|year=1998|publisher=Superintendent, Government Press}}&lt;/ref&gt; மற்ற முகனையான தொழிற்சாலைகளில் ஒன்றாகத் ''திருச்சி வடிமனை மற்றும் வேதிப்பொருள் வரையறுக்கப்பட்டது'' (TDCL) 1966ஆம் ஆண்டில் செந்தண்ணீர்புரத்தில் நிறுவப்பட்டது.&lt;ref name=&quot;madrasdistp553&quot;&gt;{{cite book|title=Madras District Gazetteers: Tiruchirappalli (pt. 1–2)|page=553|year=1998|publisher=Superintendent, Government Press}}&lt;/ref&gt; இங்கு தெளிந்த சாராவி,&lt;ref name=&quot;madrasdistp553&quot;/&gt; அசிடால்டெஹைடு,&lt;ref name=&quot;madrasdistp553&quot; /&gt; [[அசிட்டிக் காடி]],&lt;ref name=&quot;madrasdistp553&quot; /&gt; அசிடிக் அன்ஹைடிரைடு&lt;ref name=&quot;narcotip56&quot;&gt;{{cite book|title=Manufacture of Narcotic Drugs, Psychotropic Substances and Their Precursors|page=56|publisher=United Nations Publications|id={{ISBN|92-1-048082-1}}, {{ISBN|978-92-1-048082-6}}}}&lt;/ref&gt; மற்றும் இதைல்  அசிடேட்டு தயாரிக்கப்படுகின்றன. தமிழ்நாட்டில் உள்ள மிகப்பெரும் தனியார்துறை வடிமனைகளில் மிகப் பெரியதான ஒன்றாக விளங்குகிறது; திசம்பர் 2005இலிருந்து நவம்பர் 2006க்கு இடையில் 13.5 மில்லியன் லிட்டர்கள் சாராயம் தயாரிக்கப்பட்டது.&lt;ref name=&quot;gov_prohibition&quot;&gt;{{cite web|title=Demand No. 37: Prohibition and Excise|publisher=Home, Prohibition and Excise Department|page=12|url=http://www.tn.gov.in/policynotes/archives/policy2007-08/pdf/prohibition_excise.pdf|year=2007–08|accessdate=2011-05-12|format=PDF}}&lt;/ref&gt; திருச்சிராப்பள்ளி இந்தியாவின் '''&quot;ஆற்றல் திறன் மற்றும் கட்டுருவாக்கத் தொழில் தலைநகரம்&quot;''' (''Energy Equipment &amp; Fabrication Capital of India'') என்று அழைக்கப்படுகிறது.
 
 இந்த மண்டலத்திலிருந்து ஏற்றுமதியாகும் மென்பொருட்களின் ஆண்டு வருமானம் {{INR}}26.21 கோடிகளாக (US$5.8 மில்லியன்) உள்ளது.&lt;ref&gt;{{cite news|title=Software exports from Tiruchi set to rise|url=http://www.thehindu.com/news/cities/Tiruchirapalli/software-exports-from-tiruchi-set-to-rise/article943525.ece|accessdate=21 November 2013|newspaper=The Hindu|date=10 December 2010|first=R.|last=Krishnamoorthy}}&lt;/ref&gt;&lt;ref name=&quot;economictimes_it&quot;&gt;{{cite news|title=Trichy: IT infrastructure to pep up property prices|url=http://economictimes.indiatimes.com/features/property/trichy-it-infrastructure-to-pep-up-property-prices/articleshow/5429292.cms|work=The Economic Times|accessdate = 11 May 2011|date=10 January 2010}}&lt;/ref&gt; திசம்பர் 9, 2010இல் {{INR}}60 கோடிகள் (US$13.5 மில்லியன்) செலவில் ''எல்காட் தகவல்தொழில்நுட்ப பூங்கா'' திறக்கப்பட்டது.&lt;ref name=&quot;thehindu_tiruchiitcomm&quot;&gt;{{cite news|title=Tiruchi IT Park commissioned|url=http://www.hindu.com/2010/12/10/stories/2010121061490100.htm|work=The Hindu|location=India|date=10 December 2010|accessdate=2011-05-11|archivedate=2010-12-12|archiveurl=https://web.archive.org/web/20101212220301/http://www.hindu.com/2010/12/10/stories/2010121061490100.htm|url-status=dead}}&lt;/ref&gt;&lt;ref name=&quot;thehindu_itpark&quot;&gt;{{cite news|title=Tiruchi gets IT park|date=10 December 2010|url=http://www.hindu.com/2010/12/10/stories/2010121059100100.htm|work=The Hindu|location=India|accessdate=2011-05-11|archivedate=2010-12-13|archiveurl=https://web.archive.org/web/20101213205513/http://www.hindu.com/2010/12/10/stories/2010121059100100.htm|url-status=dead}}&lt;/ref&gt; தமிழ்நாடு மின்னணுக்கழகம் வரையறையால் {{Convert|59.74|ha|acre}} பரப்பளவில் அமைக்கப்பட்டுள்ள இந்த வளாகம் [[சிறப்பு பொருளாதார மண்டலம்|சிறப்பு பொருளாதார மண்டலமாக]] விளங்குகிறது.&lt;ref name=&quot;thehindu_itpark&quot; /&gt;&lt;ref name=&quot;thehindu_20080719&quot;&gt;{{cite news |url=http://www.hindu.com/2008/07/19/stories/2008071953840500.htm |title=Tamil Nadu / Tiruchi News : ELCOT to build 50,000 sq.ft. of office space in Tier II cities |work=The Hindu |location=India |date=19 July 2008 |accessdate=2011-05-11 |archivedate=2008-07-25 |archiveurl=https://web.archive.org/web/20080725093732/http://www.hindu.com/2008/07/19/stories/2008071953840500.htm |url-status=dead }}&lt;/ref&gt; இந்தியாவின் முன்னணி மென்பொருள் நிறுவனங்களில் ஒன்றான [[இன்ஃபோசிஸ்]], திருச்சிராப்பள்ளியில் தனது செயற்பாட்டைத் துவக்கத் திட்டமிட்டுள்ளது.&lt;ref name=&quot;economictimes_20090728&quot;&gt;{{cite news|url=http://articles.economictimes.indiatimes.com/2009-07-28/news/27653675_1_tier-ii-software-major-infosys-managing-director-s-gopalakrishnan |title=Infosys eyeing Tier-II cities for expansion |work=The Economic Times |location=India |date=28 July 2009|accessdate = 2011-05-11}}&lt;/ref&gt; திருச்சி மண்டல ஜி.எஸ்.டி. ஆணையர் ஜே.எம்.கென்னடி.&lt;ref name=&quot;மாலைமலர்&quot;&gt;{{cite web | url=http://www.maalaimalar.com/News/TopNews/2017/06/29162406/1093610/GST-Commissioner-says-everyone-will-get-profit-by.vpf | date=29 சூன் 2017 | accessdate=30 சூன் 2017 | title=காப்பகப்படுத்தப்பட்ட நகல் | archive-date=2017-06-29 | archive-url=https://web.archive.org/web/20170629144249/http://www.maalaimalar.com/News/TopNews/2017/06/29162406/1093610/GST-Commissioner-says-everyone-will-get-profit-by.vpf | url-status= }}&lt;/ref&gt;
 
 ==திருச்சி மாநகர் பகுதிகள்==
 
 * [[உறையூர்]]
 * [[தில்லைநகர்]]
 * [[தென்னுர்]]
 * புத்தூர்
 * ஸ்ரீனிவாசன் நகர்
 * உய்யகொண்டான் திருமலை
 * நாச்சிக்குறிச்சி
 * ராஜகாலனி
 * [[கருமண்டபம்]]
 * பிராட்டியூர் 
 * [[எடமலைப்பட்டிபுதூர்]]
 * [[பஞ்சப்பூர்]]
 * கேகே. நகர் 
 * [[மன்னார்புரம்]]
 * சுப்பிரமணியப்புரம் 
 * கல்லுக்குழி
 * [[பொன்மலைப்பட்டி]]
 * [[பொன்மலை கோல்டன் ராக்]]
 * டிவிஎஸ் டோல்கேட் l
 * கீழ் கல்கண்டார்கோட்டை
 * [[மேல் கல்கண்டார்கோட்டை]]
 * காட்டுர்
 * [[திருவெறும்பூர்]]
 * [[துவாக்குடி]]
 * குவளைக்குடி
 * [[அரியாமங்கலம்]]
 * [[பால்ண்ணை]]
 * [[வராகனேரி]]
 * சங்கிலியாண்டப்புரம் 
 * [[காந்தி மார்க்கெட்]]
 * [[பாலக்கரை]]
 * அண்ணாநகர்
 * [[மேலப்புதூர்]]
 * ஒத்தக்கடை
 * [[பீமநகர்]]
 * ஆழ்வார் தோப்பு 
 * மரக்கடை 
 * [[கோட்டை]]
 * [[பெரியக்கடைவீதி]]
 * [[சின்னக்கடைவீதி]]
 *  [[NSBசாலை]]
 * தேவதானம்
 * [[கோஹினுர்]]
 * [[கீழசிந்தமணி]]
 * [[மேலசிந்தமணி]]
 * முத்தரசாநல்லூர்
 * கம்பரசம்ம்பேட்டை 
 * [[ஸ்ரீரங்கம்]]
 * [[திருவானைக்கோவில்]]
 * திருவளர்ச்சோலை
 * [[மாம்பழச்சாலை]]
 * [[அம்மாமண்டபம்]]
 * No 1 டோல்கேட்
 * பிச்சைண்டவர்கோவில்
 * உத்தமர்கோவில்
 * தாளக்குடி
 * கூத்தூர்
 *
 *
 
 ==திருச்சி புறநகர் பகுதிகள்==
 *  [[துவாக்குடி]]
 * மாத்தூர்
 * [[சோமரசம்பேட்டை]]
 * அல்லித்துறை
 * [[குமார வயலூர்]]
 * [[இனம் குளத்தூர்|இனாம்குளத்தூர்]]
 * [[நவலூர் குட்டப்பட்டு]] 
 * [[சமயபுரம்]]
 * [[மாடக்குடி ஊராட்சி|மாடக்குடி]]
 * [[மண்ணச்சநல்லூர்]]
 * நொச்சியம்
 * [[தாளக்குடி ஊராட்சி|தாளக்குடி]]
 * [[வாளாடி ஊராட்சி|வாளாடி]]
 * சர்கார்பாளையம்
 * [[கல்லணை]]
 * [[குழுமணி ஊராட்சி|குழுமணி]]
 * [[ஜீயபுரம்]]
 * [[திருப்பராய்த்துறை]]
 * [[பெருகமணி ஊராட்சி|பெருகமணி]]
 * [[சிறுகமணி]]
 * [[பெட்டவாய்த்தலை]]
 * காவல்கரன்பட்டி 
 * [[பெட்டவாய்த்தலை]]
 * [[சிறுகனூர் ஊராட்சி|சிறுகனூர்]] 
 * [[புள்ளம்பாடி]]
 
 == சாலை மற்றும் பேருந்து போக்குவரத்து ==
 [[படிமம்:Bus-long-townbus-Trichy-Tamil Nadu.jpg|மாநகர தொடர்பேருந்து, சத்திரம்பேருந்து நிலையம்|thumb|right]]
 [[File:Trichy Central Bus Station.jpg|மத்திய பேருந்து நிலையம்|thumb|right]]
 திருச்சிராப்பள்ளி சாலை, தொடருந்து, வான்வழியாக, இந்தியாவின் பல நகரங்களுடன் இணைக்கப்பட்டுள்ளது.
 
 === பேருந்து போக்குவரத்து ===
 * தேசிய நெடுஞ்சாலைகள் [[தேசிய நெடுஞ்சாலை 45 (இந்தியா)|தேநெ 45]], தேநெ 45பி, [[தேசிய நெடுஞ்சாலை 67 (இந்தியா)|தேநெ 67]], தேநெ 210, தேநெ 227, ஆகியவை இதன் வழியாகச் செல்கின்றன. [[அரசுப் போக்குவரத்துக் கழகம் (கும்பகோணம்) கோட்டம்|தமிழ்நாடு அரசுப் போக்குவரத்துக் கழகம் - கும்பகோணம்]] தலைமையிடமாக கொண்ட [[திருச்சி|திருச்சி மண்டலம்]] சார்ந்த பேருந்துகள் என தமிழ்நாட்டின் மையபகுதியான [[திருச்சி]]யிலிருந்து '''&quot;மத்திய பேருந்து நிலையம்&quot;''' மற்றும் '''&quot;சத்திரம் பேருந்து நிலையம்&quot;''' ஆகிய இரண்டு பெரிய பேருந்து நிலையங்களில் இருந்து பல ஊர்களுக்கு பல அண்டை மாநிலங்களுக்கும் பேருந்து போக்குவரத்து நடைபெறுகின்றது.
 * திருச்சி - திருவண்ணாமலை வழியாக திருப்பதி செல்லலாம்
 * திருச்சி - [[தஞ்சாவூர்]] மார்க்கமாக : [[திருவெறும்பூர்]], [[துவாக்குடி]], [[செங்கிப்பட்டி]], [[வல்லம்]], [[தஞ்சாவூர்]], [[பாபநாசம் (தஞ்சாவூர் மாவட்டம்)|பாபநாசம்]], [[சுவாமிமலை]], [[கும்பகோணம்]], [[ஆடுதுறை]], [[மயிலாடுதுறை]],  [[வைத்தீஸ்வரன்கோயில்]], [[சீர்காழி]], [[கொள்ளிடம் ஊராட்சி ஒன்றியம்|கொள்ளிடம்]], [[சிதம்பரம் (நகரம்)|சிதம்பரம்]], [[கடலூர்]], [[புதுச்சேரி (நகரம்)|புதுச்சேரி]], [[பூம்புகார்]], [[திருக்கடையூர்]], [[பேரளம்]], [[திருநள்ளாறு|திருநள்ளார்]], [[காரைக்கால்]], [[ஒரத்தநாடு]], [[பட்டுக்கோட்டை]], [[மனோரா]], [[நீடாமங்கலம்]], [[திருவாரூர்]], [[மன்னார்குடி]], [[திருத்துறைப்பூண்டி]], [[எட்டுக்குடி ஊராட்சி|எட்டுக்குடி]], [[நாகப்பட்டினம்]], [[திருக்குவளை]], [[சிக்கல் (ஊர்)|சிக்கல்]], [[நாகூர்]], [[வேளாங்கண்ணி]], [[வேதாரண்யம்]] வரை
 * திருச்சி - [[கரூர்]]–'''[[கோவை]]''' மார்க்கமாக : [[பெட்டவாய்த்தலை]], [[குளித்தலை]], [[மாயனூர்]], [[புலியூர்]], [[கரூர்]], [[காங்கேயம்]], [[ஈரோடு]], [[தாராபுரம்]], [[திருப்பூர்]], [[பல்லடம்]], [[கோயம்புத்தூர்]], [[கோபிச்செட்டிப்பாளையம்]], [[சத்தியமங்கலம்]], [[மேட்டுப்பாளையம்]], [[ஊட்டி]] மற்றும் [[பாலக்காடு]] வரை ([[கேரளம்|கேரளா]] - [[கர்நாடகம்|கருநாடகா]] [[மாநிலம்|மாநிலங்கள்]]) வரை 
 * திருச்சி - [[பெரம்பலூர்]]–'''[[சென்னை]]''' மார்க்கமாக : [[சமயபுரம்]], [[பாடாலூர் ஊராட்சி|பாடாலூர்]], [[சிறுவாச்சூர்]], [[பெரம்பலூர்]], [[இலப்பைகுடிக்காடு]], [[திட்டக்குடி]], [[சின்னசேலம்]], [[கள்ளக்குறிச்சி]], [[உளுந்தூர்ப்பேட்டை]], [[திருக்கோவிலூர்]], [[விழுப்புரம்]], [[திருவண்ணாமலை]], [[திண்டிவனம்]], [[மேல்மருவத்தூர்]], [[செங்கல்பட்டு]], [[சென்னை]], [[காஞ்சிபுரம்]], [[திருத்தணி]], [[திருவள்ளூர்]], [[செஞ்சி]], [[ஆரணி]], [[வேலூர்]], [[காட்பாடி]], [[சித்தூர்]], [[திருப்பதி]] மற்றும் [[ஆந்திர பிரதேசம்]], [[வட இந்தியா|வட மாநிலங்கள்]] வரை
 * திருச்சி - [[மதுரை]] மார்க்கமாக : [[விராலிமலை|இலுப்பூர்,அன்னவாசல், புதுக்கோட்டை,பொன்னமராவதி,]]&lt;nowiki/&gt;விராலிமலை, [[துவரங்குறிச்சி]], [[நத்தம்]], [[மேலூர்]], [[மதுரை]], [[சிவகங்கை]], [[விருதுநகர்]], [[சிவகாசி]], [[ஸ்ரீவில்லிபுத்தூர்]], [[குற்றாலம்]], [[திருநெல்வேலி]], [[பாளையங்கோட்டை]], [[தூத்துக்குடி]], [[திருச்செந்தூர்]], [[நாகர்கோயில்]], [[கன்னியாகுமரி]] மற்றும் [[திருவனந்தபுரம்]] வரை
 * திருச்சி - [[மணப்பாறை]]–'''[[திண்டுக்கல்]]''' மார்க்கமாக : [[மணப்பாறை]], [[சிங்கம்புனரி]], [[வையம்பட்டி]], [[அய்யலூர்]], [[திண்டுக்கல்]], [[வத்தலகுண்டு]], [[கொடைக்கானல்]], [[பெரியகுளம்]], [[தேனி]], [[போடி]], [[கம்பம்]], [[ஒட்டன்சத்திரம்]], [[பழநி]], [[உடுமலைபேட்டை]], [[பொள்ளாச்சி]], [[கொச்சிக் கோட்டை கடற்கரை|கொச்சின் துறைமுகம்]] [[கேரளம்|கேரளா மாநிலங்கள்]] வரை
 * திருச்சி - [[அரியலூர்]]–'''[[பாண்டிச்சேரி]]''' மார்க்கமாக : [[லால்குடி]], [[புள்ளம்பாடி]], [[கல்லக்குடி]], [[கீழப்பழுர் ஊராட்சி|கீழப்பழுர்]], [[அரியலூர்]], [[செந்துறை]], [[பெண்ணாடம்]], [[விருத்தாசலம்]], [[ஜெயங்கொண்டம்]], [[ஆண்டிமடம்]], [[ஸ்ரீமுஷ்ணம்]], [[கங்கைகொண்ட சோழபுரம்]], [[காட்டு மன்னார் கோவில்]], [[சிதம்பரம் (நகரம்)|சிதம்பரம்]], [[பிச்சாவரம்]], [[பண்ருட்டி (கடலூர்)|பண்ருட்டி]], [[நெய்வேலி]], [[வடலூர்]], [[குறிஞ்சிப்பாடி]], [[கடலூர்]], [[புதுச்சேரி (நகரம்)|புதுச்சேரி]] வரை
 * திருச்சி - [[புதுக்கோட்டை]]–'''[[இராமேஸ்வரம்]]''' மார்க்கமாக : [[கீரனூர் (புதுக்கோட்டை)|இலுப்பூர்,கீரனூர்]], [[புதுக்கோட்டை]], [[அறந்தாங்கி]], [[திருமயம்]], [[காரைக்குடி]], [[தேவகோட்டை]], [[திருப்பத்தூர் (சிவகங்கை மாவட்டம்)|திருப்பத்தூர்]], [[சிவகங்கை]], [[மானாமதுரை]], [[பரமக்குடி]], [[திருவாடானை]], [[தொண்டி]], [[தேவிபட்டினம்]], [[இராமநாதபுரம்]], [[ஏர்வாடி]], [[இராமேஸ்வரம்]] வரை
 * திருச்சி - [[முசிறி]]–'''[[சேலம்]]''' மார்க்கமாக : [[குணசீலம்]], [[முசிறி (திருச்சி மாவட்டம்)|முசிறி]], [[தொட்டியம்]], [[காட்டுப்புத்தூர்|காட்டுபுத்தூர்]], [[தாத்தையங்கார்ப்பேட்டை]], [[மேட்டுபாளையம் ஊராட்சி|மேட்டுப்பாளையம்]], [[புளியஞ்சோலை]], [[நாமக்கல்]], [[கொல்லி மலை]], [[திருச்செங்கோடு]], [[ஈரோடு]], [[கோபிசெட்டிபாளையம்]], [[சத்தியமங்கலம்]], [[இராசிபுரம்]], [[எடப்பாடி]], [[சங்ககிரி]], [[சேலம்]], [[ஏற்காடு]], [[மேட்டூர்]], [[தருமபுரி]], [[ஒகேனக்கல்]], [[கிருஷ்ணகிரி]], [[ஒசூர்]] மற்றும் [[பெங்களூரு]] [[கர்நாடகம்|கர்நாடகா மாநிலம்]], [[வட இந்தியா|வட மாநிலங்கள்]] வரை
 * திருச்சி - [[துறையூர்]] மார்க்கமாக : [[மண்ணச்சநல்லூர்]], [[புலிவலம்]], [[துறையூர்]], [[தம்மம்பட்டி]], [[பச்சைமலை]], [[ஆத்தூர்]], [[வாழப்பாடி]], [[சேலம்]] வரை
 * திருச்சி - [[தோகைமலை ஊராட்சி|தோகைமலை]] மார்க்கமாக : [[உய்யக்கொண்டான் மலை உஜ்ஜீவநாதர் கோயில்|உய்யகொண்டான் திருமலை]], [[சோமரசம்பேட்டை]], [[வயலூர் சுப்பிரமணிய சுவாமி திருக்கோயில்|வயலூர்]], [[நங்கவரம்]], [[அல்லிதுரை ஊராட்சி|அல்லிதுறை]], [[நெய்தலூர் ஊராட்சி|நெய்தலூர் காலணி]], [[ராச்சாண்டார் திருமலை]], காவக்காரன்பட்டி, [[தோகைமலை ஊராட்சி ஒன்றியம்|தோகைமலை]], [[வீரப்பூர்]], படுகளம், [[பாலவிடுதி ஊராட்சி|பாலவிடுதி]], [[கடவூர் ஊராட்சி|கடவூர்]], [[அய்யர்மலை]], [[பாளையம் (திண்டுக்கல்)|பாளையம்]], [[தரகம்பட்டி ஊராட்சி|தரகம்பட்டி]], குவாரிசைட், [[குஜிலியம்பாறை]], [[அரவக்குறிச்சி]] வரை
 * திருச்சி - [[திருவையாறு]] மார்க்கமாக : [[திருவானைக்கோவில்]], [[கல்லணை]], [[பூண்டி மாதா கோவில்|பூண்டி மாதாகோயில்]], [[திருக்காட்டுப்பள்ளி]], [[திருவையாறு]], [[திங்களுர்]], [[கும்பகோணம்]], ,கதிராமங்கலம்,அஞ்சாறுவார்த்தலை,மயிலாடுதுறை,மனக்குடி,கருவி[[பூம்புகார்]] வரை 
 * திருச்சியில் இருந்து தினசரியாக இந்த ஊர்களுக்கு பேருந்துகள் இயங்குகின்றன.
 
 {| class=&quot;wikitable&quot;
 |-
 ! தேசிய நெடுஞ்சாலை எண் !! தொடங்கும் இடம் !! முடியும் இடம்
 |-
 | தேசிய நெடுஞ்சாலை 45 || [[சென்னை]] - திருச்சி ||  திருச்சி - [[திண்டுக்கல்]]
 |-
 | தேசிய நெடுஞ்சாலை 45பி || திருச்சி || [[தூத்துக்குடி]]
 |-
 | தேசிய நெடுஞ்சாலை 67 || [[நாகப்பட்டினம்]] || குண்டல்பேட்
 |-
 | தேசிய நெடுஞ்சாலை 210 || திருச்சி || [[இராமநாதபுரம்]]
 |-
 | தேசிய நெடுஞ்சாலை 227 || திருச்சி || [[சிதம்பரம்]]
 |}
 
 ==== பேருந்து நிலையங்கள் ====
 [[File:Layout of CBS, Trichy.jpg|thumb|right|மத்திய பேருந்து நிலையத்தின் வரைப்படம்]]
 திருச்சியைப் பொறுத்தவரை நான்கு  முக்கிய நிலையைங்களாகவும் நகரில் பகுதி பேருந்து நிலையங்களாக மூன்று பேருந்து நிலையங்கள் உள்ளன.
 # '''[[பஞ்சாப்பூர் ஒருங்கிணைந்த பேருந்து முனையம்,திருச்சிராப்பள்ளி|பஞ்சாப்பூர் ஒருங்கிணைந்த பேருந்து முனையம்]]:'''இது நகரின் தென் மேற்கு பகுதியில் அமைந்துள்ளது தமிழகத்தில் மிகப்பெரிய பேருந்து முனையம் ஆகும்.
 # '''[[மத்திய பேருந்து நிலையம், திருச்சிராப்பள்ளி|மத்திய பேருந்து நிலையம்]]:''' இது நகரின் தென்மேற்கு பகுதியில் இருக்கிறது. வெளியூர் பேருந்துகளுக்கு மட்டும் வந்து புறப்படும் இடமாக உள்ளது. இங்கிருந்து தொடருந்து நிலையமும், விமான நிலையமும் அருகாமையில் இருக்கிறது.
 # '''[[சத்திரம் பேருந்து நிலையம், திருச்சிராப்பள்ளி|சத்திரம் பேருந்து நிலையம்]]:''' இது நகரின் வடகிழக்குப் பகுதியில் இருக்கிறது. [[திருச்சி மலைக் கோட்டை|மலைக்கோட்டைக்கு]] அருகிலுள்ளது. இது நகரப்பேருந்துகள் மற்றும் திருச்சியின் அண்மையில் உள்ள முக்கிய நகரங்களுக்கு பேருந்துகள் இயங்கும் நிலையமாக உள்ளது.
 # '''[[புதிய பேருந்து நிலையம் ஸ்ரீரங்கம்,திருச்சிராப்பள்ளி|புதிய பேருந்து நிலையம் ஸ்ரீரங்கம்]]:'''இது வடக்கு பகுதியில் ஸ்ரீரங்கம் தீவில் இருக்கிறது [[ஸ்ரீரங்கம்]]அருகில் உள்ளது
 # '''[[No. 1 Tollgate பேருந்து நிலையம்]]'''  :இது கொள்ளிடம் ஆறு வடக்கு பகுதியில் அமைந்துள்ளது.
 #'''[[திருவெறும்பூர் பேருந்து நிலையம்]]:'''இது மாநகர் கிழக்கு பகுதியில் அரியமங்கலம் கோட்டம் தஞ்சை தேசிய நெடுஞ்சாலையில் அமைந்துள்ளது.
 #'''[[K.K நகர் பேருந்து நிலையம்]]:''' இது நகரின் தெற்கு பகுதியின் kk நகரில் அமைந்துள்ளது.
 #'''[[சமயபுரம் பேருந்து நிலையம்]]'''  : நகரின் வடக்கு பகுதியில் அமைந்துள்ளது.
 #'''[[துவாக்குடி பேருந்து நிலையம்]]:'''இது நகரின் கிழக்கு பகுதியில் துவாக்குடி தஞ்சை தேசிய நெடுஞ்சாலையில் அமைந்துள்ளது.
 
 ==== பேருந்து போக்குவரத்து மாற்றம் ====
 திருச்சி தமிழகத்தின் நடுவில் இருப்பதாலும், மக்கள் தொகை நெருக்கத்தாலும், நகரத்துக்குள் போக்குவரத்து நெரிசலைக் குறைப்பதற்காகவும் பேருந்து நிறுத்தங்களில் மாற்றங்கள் ஏற்படுத்தப்பட்டன. அதன்படி, நகரின் தெற்கிலும், மேற்கிலும் இருந்து வரும் வெளியூர் பேருந்துகள், பெரும்பாலும் மத்திய பேருந்து நிலையத்தை இறுதி நிறுத்தமாகக் கொள்வர்.
 
 அதேபோல, வடக்கிலும், கிழக்கிலும் இருந்து வரும் வெளியூர் பேருந்துகள், சத்திரம் பேருந்து நிலையத்தை இறுதி நிறுத்தமாகக் கொள்வர். இங்கிருந்து மத்திய பேருந்து நிலையத்திற்கு ஏறத்தாழ 7 [[கிலோமீட்டர்|கி.மீ.]] இடைவெளியுள்ளது. இதனால் மாநகரில், வெளியூர் பேருந்துகளால் ஏற்படும் நெரிசல் குறைகிறது.
 
 பயணிகளும், தங்களது பயணங்களுக்கு ஏற்ப, பேருந்தைத் தேர்ந்தெடுத்துச் செல்வர். இதனால், நகரில் வாழ்வோருக்கு, இயல்பு வாழ்க்கை எளிதாக்கப்படுகிறது. வெளியூர் பயணிகள் மட்டும், ஒரு பேருந்து நிலையத்தில் இறங்கி, மற்றொரு நிலையத்திற்கு வர, சில நேரங்களில் சிரமப்படும் வாய்ப்பு இருக்கிறது. நகருக்குள் புகுந்து, மறுபக்கம் பயணிக்கும் பேருந்துகள் புறவழிச்சாலையையும் பயன்படுத்த, நெடுஞ்சாலை வசதி செய்யப்பட்டுள்ளது. இப்புறவழிச்சாலையை [[சுமையுந்து]]களும் (''lorry''), [[மகிழுந்து]]களும் (''cars'') அதிகம் பயன்படுத்துகின்றன. இதனால், தமிழகத்தின் இருகோடியிலிருந்து பயணிக்கும் வண்டிகள், இடரில்லாமல் பயணிக்கின்றன.
 
 == இரயில் போக்குவரத்து ==
 [[படிமம்:Tiruchirapalli junction.JPG|250px|thumbnail|வலது|திருச்சிராப்பள்ளி தொடருந்து நிலைய சந்திப்பு]]
 * [[திருச்சிராப்பள்ளி சந்திப்பு]], [[தென்னக இரயில்வே]]யின் ஆறு கோட்டங்களில் ஒன்றாகும். 1868ம் ஆண்டு திருச்சிராப்பள்ளிக்கும் நாகப்பட்டினத்திற்கும் இடையே தொடருந்து சேவை தொடங்கியது. சென்னை, கோயம்புத்தூர், திருவனந்தபுரம், கன்னியாகுமரி, தஞ்சாவூர், மதுரை, விழுப்புரம், ஸ்ரீ கங்காநகர், ஜோத்பூர், ஜெய்ப்பூர், புனே, வதோதரா, நாக்பூர், மயிலாடுதுறை, நாகப்பட்டினம், காரைக்கால், கடலூர், சிதம்பரம், ஈரோடு, ஹைதராபாத், கொல்லம், இராமேசுவரம், தென்காசி, திருப்பதி, பெங்களூரு, மைசூரு, மங்களூர், கொல்கத்தா, குவஹாத்தி, கொச்சி ஆகிய இடங்களுக்குத் தொடருந்து வசதி உண்டு.
 * தினசரி 37+ தொடருந்துகள் திருச்சியில் இருந்து மாநில தலைநகர் சென்னைக்குச் செல்கின்றன. இதில் சோழன், மலைக்கோட்டை விரைவு வண்டி  திருச்சியில் இருந்து புறப்படுகிறது, மற்ற தொடருந்துகளான வைகை, பல்லவன், அனந்தபுரி மற்றும் இதர இரயில்கள் தென் மாவட்டங்களில் இருந்து திருச்சி வழியாகச் சென்னை செல்லுகின்றன.
 * மேலும் [[திருச்சி]]யில் வழிதட இரயில் நிலையங்களான [[திருச்சிராப்பள்ளி டவுன் ரயில் நிலையம்|திருச்சிராப்பள்ளி டவுன் ஸ்டேசன்]] [[திருச்சி]]–[[சென்னை]] வழிதடத்தில் [[திருச்சி மலைக் கோட்டை|மலைக்கோட்டை]] திருக்கோயிலுக்கு அருகில் அமைந்துள்ள இந்த இரயில் நிலையத்தில் திருச்சி மாநகர பகுதியில் உள்ளவர்கள் [[சென்னை]], [[விழுப்புரம்]], [[கடலூர்]], ([[திருப்பதி]]/[[ஆந்திரப் பிரதேசம்|ஆந்திர மாநிலம்]]), [[தெலுங்கானா|தெலுங்கானா மாநிலம்]]) மற்றும் வடமாநில மார்க்கமாக செல்லும் இரயில்களில் பயணிகள் செல்வதற்கு வசதியாக இந்த '''டவுன் ஸ்டேசன்''' இரயில் நிலையம் உதவியாக உள்ளது.
 * திருச்சியின் மைய நகர பகுதியில் [[திருச்சி]]–[[கோவை|ஈரோடு]] வழித்தடத்தில் [[திருச்சிராப்பள்ளி கோட்டை தொடருந்து நிலையம்|திருச்சிராப்பள்ளி கோட்டை ஸ்டேசன்]] [[ஈரோடு]], [[கோவை]], [[சேலம்]] மற்றும் [[கேரள மாநிலம்]], [[கர்நாடக மாநிலம்]] மற்றும் பிற வடமாநில மார்க்கமாக செல்லும் இரயில்களில் பயணிகள் செல்வதற்கு இந்த '''கோட்டை ஸ்டேசன்''' உதவியாக.
 உள்ளது.
 
 ======சந்திப்பு மற்றும் தொடர்ந்து நிலையங்கள்======
 
 #[[திருச்சிராப்பள்ளி சந்திப்பு தொடருந்து நிலையம்|திருச்சிராப்பள்ளி சந்திப்பு]].
 #[[திருவரங்கம் தொடருந்து நிலையம்]],
 #[[திருச்சிராப்பள்ளி கோட்டை தொடருந்து நிலையம்|கோட்டை தொடருந்து நிலையம்]],
 #[[திருவெறும்பூர்|திருவெறும்பூர் தொடருந்து நிலையம்]],
 #[[பொன்மலை தொடருந்து நிலையம்]],
 #[[டவுன் தொடருந்து நிலையம்]],
 #[[பாலக்கரை தொடருந்து நிலையம்]]
 #[[மஞ்சுத்திடல் தொடருந்து நிலையம்]]
 #[[முத்தாரசாநல்லூர் தொடருந்து நிலையம்]]
 #[[பூங்குடி தொடருந்து நிலையம்]] 
 #[[பிச்சாண்டவர் கோவில் தொடருந்து நிலையம்]]
 #[[உத்தமர் கோவில் தொடருந்து நிலையம்]]
 #[[வாளடி தொடருந்து நிலையம்]]
 #[[குமாரமங்கலம் தொடருந்து நிலையம்]]
 
 '''தினசரி திருச்சிராப்பள்ளி வழியாகச் செல்லும் தொடருந்துகள்''' &lt;ref&gt;[http://www.cleartrip.com/trains/results?from_city=Trichy&amp;to_city=Chennai&amp;class=2S&amp;date=5-9-2012&amp;adults=1&amp;children=0&amp;male_seniors=0&amp;female_seniors=0 கிளியர்டிரிப் தளத்தில் திருச்சி வழியாக சென்னை செல்லும் தொடருந்துகளை தேடும் பொழுது கிடைத்த முடிவுகள்]{{Dead link|date=பிப்ரவரி 2023 |bot=InternetArchiveBot }}&lt;/ref&gt;.
 {| class=&quot;wikitable&quot;
 |-
 ! வண்டியின் பெயர் !! வண்டி எண் !! புறப்படும் இடம்!! சேருமிடம்
 |-
 | அனந்தபுரி விரைவு வண்டி || 16724 || திருவனந்தபுரம் || சென்னை
 |-
 | பொதிகை அதிவிரைவு வண்டி || 12662 || செங்கோட்டை || சென்னை
 |-
 | பல்லவன் அதிவிரைவு வண்டி || 12606 || காரைக்குடி || சென்னை
 |-
 | வைகை அதிவிரைவு வண்டி || 12636 || மதுரை || சென்னை
 |-
 | சேது அதிவிரைவு வண்டி || 22661 || இராமேசுவரம் || சென்னை
 |-
 | குருவாயூர் விரைவு வண்டி|| 16127|| குருவாயூர் || சென்னை
 |-
 | சென்னை || 06804 || செங்கோட்டை || சென்னை
 |-
 | மங்களூர் விரைவு வண்டி || 16159 || மங்களூர் || சென்னை
 |-
 | மலைக்கோட்டை அதிவிரைவு வண்டி || 16178 || '''திருச்சிராப்பள்ளி''' || சென்னை
 |-
 | திருச்செந்தூர் விரைவு வண்டி || 16736 || திருச்செந்தூர் || சென்னை
 |-
 | நெல்லை அதிவேக விரைவு வண்டி || 12632 || திருநெல்வேலி || சென்னை
 |-
 | கன்னியாகுமரி அதிவேக விரைவு வண்டி || 12634 || கன்னியாகுமரி || சென்னை
 |-
 | முத்துநகர் (பேர்ல் சிட்டி) அதிவேக விரைவு வண்டி || 12694 || தூத்துக்குடி || சென்னை
 |-
 | சோழன் விரைவு வண்டி || 16854 || '''திருச்சிராப்பள்ளி''' || சென்னை
 |-
 | பாண்டியன் அதிவேக விரைவு வண்டி || 12638 || மதுரை || சென்னை
 |-
 | போட்மெயில் விரைவு வண்டி || 16851 || இராமேசுவரம் || சென்னை
 |-
 | மைசூர் விரைவு வண்டி || 16231 || மயிலாடுதுறை || மைசூர்
 |}
 மைசூர் விரைவு வண்டி சேலம், ஓசூர், பெங்களூரு வழியாக மைசூரை அடைகிறது.
 
 == விமான போக்குவரத்து ==
 [[படிமம்:Trichy International Airport.JPG|200px|thumbnail|வலது|திருச்சி வானூர்தி நிலையம்]]
 
 [[திருச்சிராப்பள்ளி பன்னாட்டு வானூர்தி நிலையம்|திருச்சிராப்பள்ளி வானூர்தி நிலையம்]] தேசிய நெடுஞ்சாலை 210ல் அமைந்துள்ளது. இவ்வானூர்தி நிலையம் தமிழகத்தில் சென்னைக்கு அடுத்து இரண்டாவது பெரியவிமான நிலையமாகும்'''''.''''' 1938ல் டாடா ஏர்லைன்சின் கராச்சி-கொழும்பு தடத்தில் செல்லும் வானூர்திகள் இங்கு நின்று சென்றன. இங்கிருந்து சென்னை மற்றும் பெங்களூருக்கு வானூர்திகள் செல்லுகின்றன. சிங்கப்பூர், கொழும்பு, துபாய், அபுதாபி, கோலாலம்பூர், தோஹா, மஸ்கட் மற்றும் குவைத் ஆகிய வெளிநாட்டு நகரங்களுக்கும் வானூர்திகள் செல்லுகின்றன.
 
 == திருத்தலங்கள் ==
 * [[திருச்சிராப்பள்ளி தாயுமானவர் கோயில்|அருள்மிகு தாயுமான சுவாமி திருக்கோயில்]], [[திருச்சி மலைக் கோட்டை]], திருச்சி
 * [[திருச்சி உச்சிப்பிள்ளையார் கோயில்|அருள்மிகு உச்சிப் பிள்ளையார் திருக்கோயில்]], மலைக்கோட்டை, திருச்சி
 * [[சமயபுரம் மாரியம்மன்|அருள்மிகு மாரியம்மன் திருக்கோயில், சமயபுரம்]]
 * அருள்மிகு அரங்கநாத சுவாமி திருக்கோயில், [[திருவரங்கம்|ஸ்ரீரங்கம்]]
 * [[வெக்காளியம்மன்|அருள்மிகு வெக்காளி அம்மன் திருக்கோயில்]], [[உறையூர்]]
 * [[திருவானைக்காவல் ஜம்புகேசுவரர் கோயில்|அருள்மிகு சம்புகேசுவரர் திருக்கோயில்]], [[திருவானைக்காவல்]]
 * [[திருவெறும்பூர் எறும்பீஸ்வரர் கோயில்]]
 * [[வயலூர் சுப்பிரமணிய சுவாமி திருக்கோயில்]]
 * அருள்மிகு பஞ்சவர்ண சுவாமி திருக்கோயில், [[உறையூர்]]
 * அருள்மிகு ஆம்ரனேஸ்வர சுவாமி திருக்கோயில், [[மாந்துறை]]
 * அருள்மிகு [[உத்தமர் கோயில்]]
 * அருள்மிகு பிரசன்ன வெங்கடாசலபதி திருக்கோயில் [[குணசீலம்]]
 * அருள்மிகு ஞீலிவனேஸ்வரர் திருக்கோயில் [[திருப்பைஞ்ஞீலி]]
 * அருள்மிகு பொன்னேஸ்வரி அம்மன் திருக்கோயில், [[பொன்மலை]]
 * அருள்மிகு மாரியம்மன் திருக்கோயில், [[பொன்மலை]]
 * அருள்மிகு காமாட்சி அம்மன் திருக்கோயில், [[பொன்மலை]]
 * [[சமயபுரம் மாரியம்மன்|அருள்மிகு மாரியம்மன் திருக்கோயில்,]]&lt;nowiki/&gt; அன்பில்
 * [[சமயபுரம் மாரியம்மன்|அருள்மிகு சுந்தரராஜ பெருமாள் திருக்கோயில்,]]&lt;nowiki/&gt; அன்பில்
 *சீரடி சாய் பாபா கோவில் (மேக்குடி)
 * மகா மாரியம்மன் திருக்கோவில், (கோப்பு).
 * அருள்மிகு உஜ்ஜிவநாத சுவாமி திருக்கோவில், (உ.கொ.திருமலை).
 * ஸ்ரீ வள்ளலாள கண்ட அய்யனார் திருக்கோவில், (குண்டூர்).
 * அருள்மிகு ஸ்ரீ மத்தியஜுனேஸ்வரர் திருக்கோவில், (தேவஸ்தானம்), பெட்டவாய்த்தலை.
 *அருள்மிகு ஸ்ரீ சோமகாளியம்மன் திருக்கோவில், (தேவஸ்தானம்), பெட்டவாய்த்தலை.
 * அருள்மிகு பிரம்மபுரீஸ்வரர் திருக்கோவில், திருப்பட்டுர்.
 
 == சுற்றுலாத் தலங்கள் ==
 * [[திருச்சி மலைக் கோட்டை|திருச்சி மலைக் கோட்டை (Rock fort, Trichy)]]
 * [[திருவரங்கம் அரங்கநாத சுவாமி கோயில்|ஸ்ரீரங்கம்]]
 * [[திருவானைக்கோவில்]]
 * [[இராணி மங்கம்மாள் கொலு மண்டபம்]]
 * [[முக்கொம்பு]]
 * [[கல்லணை]]
 * [[வயலூர் சுப்பிரமணிய சுவாமி திருக்கோயில்]]
 * [[முதலாம் உலகப்போர் நினைவுச்சின்னம் (திருச்சிராப்பள்ளி)]]
 * [[ஸ்ரீரங்கம் வண்ணத்துப்பூச்சி பூங்கா|வண்ணத்துப்பூச்சி பூங்கா]]
 * [[சமயபுரம் மாரியம்மன் கோயில்]] 
 * [[அண்ணா அறிவியல் மையம், திருச்சிராப்பள்ளி|கோளரங்கம்]]
 * மான் பூங்கா
 * இரயில்வே அருங்காட்சியகம்
 * இப்ராகிம் பார்க் பூங்கா
 
 ==ஆடிப்பெருக்கு விழா==
 ஒவ்வோர் ஆண்டும் தமிழ் மாதமான ஆடி மாதம், 18 ஆம் திகதி அன்று, காவிரி நதிக் கரையோரம், குடும்பப் பெண்கள், சுமங்கலிப் பெண்கள், புதிதாகத் திருமணமான பெண்கள், திருமணமாகாத கன்னிப் பெண்கள், தங்கள் குடும்பங்களுடன் சென்று பூசைகள் செய்து, தங்களின் குடும்ப நலனுக்காக காவிரித்தாயை வணங்குவர். பூசைக்காக, மலர் மாலை, பச்சரிசி, ஊதுபத்தி, சாம்பிராணி, மஞ்சள், குங்குமம், தேங்காய், வெற்றிலை, பாக்கு, பழங்கள், வெல்லம், மஞ்சள் கயிறு, கற்பூரம், விபூதி, சந்தனம், நாணயங்கள், நறுமணப் பூக்கள், தேன், பச்சரிசி மாவு, பன்னீர், வாழை இலை போன்றவற்றைக் கொண்டு சென்று, காவிரித்தாய்க்குப் படைத்து பண்டிகையாகக் கொண்டாடுவது வழக்கம். நதிக்கரையில், சர்க்கரைப் பொங்கல் மற்றும் வெண்பொங்கல் தயார் செய்து படையல் செய்து இறையருள் பெற வேண்டுகின்றனர்.
 
 == கல்வி ==
 === கல்லூரிகள் ===
 [[படிமம்:National Institute of Technology, Trichy.jpg|thumb|துவாக்குடியில் உள்ள  [[தேசிய தொழில்நுட்பக் கழகம், திருச்சி|திருச்சி தேசிய தொழிற்நுட்பக் கழகம்]]]]
 திருச்சியிலும் அதன் புறநகர் பகுதியிலுமாக 30க்கும் மேற்பட்ட கல்லூரிகள் உள்ளன.
 
 ==== கலை மற்றும் அறிவியல் கல்லூரிகள் ====
 * [[பாவேந்தர் பாரதிதாசன் கலை அறிவியல் கல்லூரி]]
 * [[தேசியக் கல்லூரி, திருச்சி]]
 * [[அரசினர் கலைக்கல்லூரி, திருச்சிராப்பள்ளி]]* [[பாரதிதாசன் பல்கலைக்கழகம்]]
 * பாரதிதாசன் பல்கலைக்கழகக் கலை மற்றும் அறிவியல் கல்லூரி, மாத்தூர்
 * [[ஜமால் முகம்மது கல்லூரி]]
 * [[பிஷப் ஹீபர் கல்லூரி, திருச்சிராப்பள்ளி|பிஷப் ஹீபர் கல்லூரி]]
 * [[புனித சூசையப்பர் கல்லூரி, திருச்சி]]
 * [[நேரு நினைவுக் கல்லூரி]], புத்தனாம்பட்டி
 * [[பெரியார் ஈ. வெ. ரா. அரசினர் கலைக்கல்லூரி]] 
 * ஆண்டவன் கலை அறிவியல் கல்லூரி
 * உருமு தனலட்சுமி கல்லூரி
 *குறிஞ்சி கலை மற்றும் அறிவியல் கல்லூரி
 * ஏ. ஏ. அரசு கலைக்கல்லூரி
 * கலைக் காவிரி கலை மற்றும் அறிவியல் கல்லூரி
 * காவேரி மகளிர் கல்லூரி
 * [[கிருத்துராஜ் கல்லூரி]]
 * [[சீத்தாலச்சுமி இராமசாமி கல்லூரி, திருச்சி]]
 * செட்டிநாடு கலை மற்றும் அறிவியல் கல்லூரி
 * தூய சிலுவைக் கல்லூரி
 * ஸ்ரீமதி இந்திராகாந்தி கல்லூரி
 * தனிநாயகம் அடிகள் இதழியல் கல்லூரி
 * M.I.E.T.  கலை மற்றும் அறிவியல் கல்லூரி
 
 ==== சட்டப் பல்கலைக்கழகம்/ கல்லூரிகள் ====
 * தமிழ் நாடு தேசிய சட்டப்பள்ளி பல்கலைக்கழகம், திண்டுக்கல் நெடுஞ்சாலை, நவலூர் குட்டப்பட்டு, திருச்சிராப்பள்ளி.
 * அரசு சட்டக் கல்லூரி, திருச்சிராப்பள்ளி.
 
 ==== பொறியியல் கல்லூரிகள் ====
 * [[அண்ணா தொழில்நுட்பப் பல்கலைக்கழகம், திருச்சிராப்பள்ளி]]
 * [[தேசிய தொழில்நுட்பக் கழகம், திருச்சி]] (முன்பு மண்டலப் பொறியியற் கல்லூரி)
 * அங்காளம்மன் பொறியியல் மற்றும் தொழில்நுட்பக் கல்லூரி, சிறுகனூர்,
 * ஜெயராம் பொறியியல் மற்றும் தொழில்நுட்பக் கல்லூரி
 * ஆக்ஸ்போர்டு பொறியியல்  கல்லூரி
 * சாரநாதன் பொறியியல் கல்லூரி
 * ஜெ.ஜெ. பொறியியல் மற்றும் தொழில்நுட்பக் கல்லூரி
 * [[திருச்சி பொறியியல் கல்லூரி]]
 * MAM  பொறியியல் கல்லூரி
 * K. ராமகிருஷ்ணன் பொறியியல் கல்லூரி ( சமயபுரம்)and technology
 * [[சிவானி பொறியியல் கல்லூரி, திருச்சிராப்பள்ளி]]
 *M.I.E.T பொறியியல் கல்லூரி.
 *SRM-TRP பொறியியல் கல்லூரி.
 
 ==== வேளாண்மைக் கல்லூரிகள் ====
 *அன்பில் தர்மலிங்கம் வேளாண்மைக் கல்லூரி மற்றும் ஆராய்ச்சி நிலையம்
 * மகளிர் தோட்டக்கலைக் கல்லூரி மற்றும் ஆராய்ச்சி நிலையம்
 *வேளாண்மை கல்வி நிறுவனம் குமுளுர், திருச்சி.
 
 ==== கல்வியியல் கல்லூரிகள் ====
 ==== மருத்துவக் கல்லூரிகள் ====
 * [[கி. ஆ. பெ. விசுவநாதம் அரசு மருத்துவக் கல்லூரி]]
 * கண்மருத்துவ நிறுவனம் (ஜோசப் கண் மருத்துவமனை)
 * எசு. ஆர். எம். மருத்துவ கல்லூரி]]
 
 === பள்ளிகள் ===
 * திருவரங்கம் ஆண்கள் மேல்நிலைப்பள்ளி
 * பாய்லர் பிளாண்ட் ஆண்கள் மேல்நிலைப்பள்ளி
 * பாய்லர் பிளாண்ட் பெண்கள் மேல்நிலைப்பள்ளி
 * ஈ.ஆர். மேனிலைப் பள்ளி
 * [[பிசப் ஈபர் மேனிலைப்பள்ளி]]
 * ஈ.வே.ரா. மேனிலைப்பள்ளி
 * புனித சிலுவைப் பெண்கள் மேனிலைப்பள்ளி
 * புனித வளனார் மேனிலைப்பள்ளி
 * தேசிய மேனிலைப்பள்ளி
 * [[கேம்பியன் மேனிலைப்பள்ளி]]
 * ஆர்.சி. மேல்நிலைப்பள்ளி
 * காஜா மியான் மேனிலைப்பள்ளி
 * திரு இருதய மேல்நிலைப்பள்ளி
 * பொன்னையா மேல்நிலைப்பள்ளி
 * இரயில்வே இருபாலர் மேல்நிலைப்பள்ளி, பொன்மலை
 * புனித ஜான் வெஸ்ட்ரி ஆங்கிலோ 
 * இந்திய மேல்நிலைப்பள்ளி   
 * அரசு ஆதி திராவிடர் நல ஆண்கள் மேல் நிலைப்பள்ளி, காட்டூர் பாப்பாகுறிச்சி
 * படைக்கல தொழிற்சாலை மேல்நிலைப்பள்ளி
 * கேந்திரிய வித்யாலயா(CBSE) (K.V-NO:1) O.F.T
 * கேந்திரிய வித்யாலயா(CBSE) (K.V-NO:2) H.A.P.P
 
 == விக்கிக்காட்சியகம் ==
 &lt;gallery&gt;
 File:22Trischinopoly Rock Temple.jpg|மலைக்கோட்டை,1913
 File:Srirangam temple tower.JPG|ஸ்ரீரங்கம் கோயில் கோபுரம்
 File:Rock Fort Temple.jpg|தெப்பக்குளம் 
 File:Rock Fortress - Tiruchirappalli - India.JPG|மலையிலிருந்து நகரம்
 File:Trichy11.jpg|பாலம்+கோட்டை
 File:Cauvery River.jpg|காவிரிப்பாலம்
 File:Trichy Kaveri Bridge.jpg|புதிய, பழைய பாலங்கள்
 File:Trichy Kaveri Rail Bridge.jpg|[[தொடருந்து]] பாலம்
 File:St.lourd'schurch.jpeg|கிறிஸ்தவ ஆலயம்
 File:Trichy central bus stand.JPG|மத்தியப் பேருந்து நிலையம்
 |தொடருந்துசந்திப்பு
 File:Trichyairportfront.jpg|[[வானூர்தி]] நிலையம்
 File:National Institute of Technology, Trichy.jpg|[[தேசிய தொழில்நுட்பக் கழகம், திருச்சி|NITT(REC)]], [[துவாக்குடி]]
 &lt;/gallery&gt;
 
 == குறிப்புகள் ==
 &lt;references group=&quot;கு&quot;/&gt;
 
 == மேற்கோள்கள் ==
 {{Reflist|3}}
 
 == வெளி இணைப்புகள் ==
 * [http://www.tagavalthalam.com/2014/01/rock-forttrichy.html திருச்சி மலைக் கோட்டை(Rock fort,Trichy) ஒரு பார்வை ]
 * [http://www.dinamalar.com/tnspl_home.asp?id=279  திருச்சி நகரின் சிறப்பு]
 * [http://temple.dinamalar.com/New.aspx?id=24  திருச்சி மலைக்கோட்டை உச்சிப்பிள்ளையார் திருக்கோயில்]
 * [http://www.trichycorporation.gov.in/  திருச்சி மாநகராட்சியின் வலைப்பக்கம்] {{Webarchive|url=https://web.archive.org/web/20100304140428/http://www.trichycorporation.gov.in/ |date=2010-03-04 }}
 * [http://www.trichy.tn.nic.in/  திருச்சி மாவட்டம் பற்றிய அரசின் வலைப்பக்கம்] {{Webarchive|url=https://web.archive.org/web/20110625205228/http://www.trichy.tn.nic.in/ |date=2011-06-25 }}
 * [http://www.tamilpayani.com/tn/tiruchirappalli/index.htm  திருச்சி மாவட்டம் பற்றிய வலைப்பக்கம்] {{Webarchive|url=https://web.archive.org/web/20051225174953/http://www.tamilpayani.com/tn/tiruchirappalli/index.htm |date=2005-12-25 }}
 
 {{திருச்சிராப்பள்ளி மாவட்டம்}}
 {{தமிழக மாநகராட்சிகள் மற்றும் நகராட்சிகள்}}
 {{Authority control}}
 [[பகுப்பு:தமிழ்நாட்டிலுள்ள மாநகரங்கள்]]

</doc>


'''ஒராதேயா''' ([[அங்கேரிய மொழி|அங்கேரியம்]] '''Nagyvárad''', [[ஜெர்மானிய மொழி|இடாய்ச்சு]] '''Großwardein''') என்பது [[உருமேனியா]] நாட்டில் [[திரான்சில்வேனியா]]விலுள்ள [[பிஹோர்]] கவுண்டியைச் சேர்ந்த ஒரு [[நகரம்|நகர]]மாகும். 2002 ஆண்டில் எடுக்கப்பட்ட மக்கள் தொகைக் கணக்கெடுப்பின்படி, நகரத்தில் 206,527 பேர் உள்ளனர். இது மாநகரசபைக்கு வெளியேயுள்ள பகுதிகளை உள்ளடக்கவில்லை. இப் பகுதிகளையும் சேர்த்தால் மொத்த நகர் சார்ந்த மக்கள் தொகை அண்ணளவாக 220,000 ஆகும். ஒரேடெயா ருமேனியாவின் மிகவும் வளம் பொருந்திய நகரங்களில் ஒன்றாகும்.
 [[படிமம்:OradeaTownHall.jpg|thumb|250px|ஒராதேய நகர மண்டபம்]]
 
 == புவியியல் ==
 
 இந்த் நகரம் ஹங்கேரிய எல்லையை அண்டி [[கிரிசுல் ரெபேடே]](Crişul Repede) நதிக்கரையில் அமைந்துள்ளது.
 
 == வரலாறு ==
 
 வராடியம் என்னும் [[லத்தீன் மொழி]]ப் பெயரில், 1113 இல் முதல் முதலாக ஒராடெயா குறிப்பிடப்படுகின்றது. இன்றும் அழிந்த நிலையில் காணப்படும் ஒராடெயா Citadel [[1241]]ல் முதல் தடவையாகக் குறிப்பிடப்படுகின்றது, எனினும் 16 ஆம் நூற்றாண்டின் பின்னரே இப்பகுதி ஒரு நகராக வளரத்தொடங்கியது. 1700 ல் [[வியன்னா]]வைச் சேர்ந்த பொறியியலாளரான [[பிரான்ஸ் அன்டன் ஹில்லெபிராண்ட்]] (Franz Anton Hillebrandt) [[பரோக்]] பாணியில் நகரத்தை வடிவமைத்தார். 1752லிருந்து ரோமன் கத்தோலிக்க தேவாலயம், பேராயர் மாளிகை, கிரிஸ்(Criş) மண்ணின் அரும்பொருட் காட்சியகம் (Muzeul Ţării Crişurilor) போன்ற முக்கிய கட்டிடங்கள் கட்டப்பட்டன.
 
 == பொருளாதாரம் ==
 
 ஹங்கேரிய எல்லையில் மேற்கு ஐரோப்பாவுக்கான நுழைவாயிலில் அமைந்திருந்ததால் ஒராடெயா நீண்ட காலமாகவே ருமேனியாவின் வளமிக்க ஒரு நகரமாக விளங்கி வந்தது. 1989 இன் பின்னர் ஒராடெயா, தொழில் வளர்ச்சி அடிப்படையிலன்றிச் சேவைத்துறை வளர்ச்சிமூலம் ஓரளவு பொருளாதாரப் புத்துயிர் பெற்றது.
 
 ஒராடெயாவின் வேலையின்மை விகிதமான 6%, ருமேனியாவின் சராசரியைவிடக் குறைவாக இருப்பினும், [[பிஹோர்]] கவுண்டியின் 2% வேலையின்மை விகிதத்துடன் ஒப்பிடுகையில் மிகவும் அதிகமாகும். [[பிஹோர்]] கவுண்டியின் 34.5% குடித்தொகையைக் கொண்ட ஒராடெயா, அக் கவுண்டியின் 63% தொழில்துறை உற்பத்திக்குக் காரணமாக உள்ளது. இதன் முக்கியமான உற்பத்திகள், தளபாடங்கள், புடவை, ஆடை உற்பத்தி, காலணிகள் மற்றும் உணவு வகைகளாகும். 
 
 [[2003]] ல் நகரின் முதலாவது பெரிய கொள்வனவு மையமான [[லோட்டஸ் சந்தை]] வர்த்தக மையம் ஒராடெயாவில் திறந்து வைக்கப்பட்டது.
 
 == இனங்கள் ==
 === வரலாறு ===
"""
```

```python
tokens = tamil_text.encode("utf-8") # raw bytes
tokens = list(map(int, tokens)) # convert to a list of integers in range 0..255 for convenience
print(tokens[0:100])
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

def merge(ids, pair, idx):
  newids = []
  i = 0
  while i < len(ids):
    if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
      newids.append(idx)
      i += 2
    else:
      newids.append(ids[i])
      i += 1
  return newids

# ---
vocab_size = 6000 # the desired final vocabulary size
num_merges = vocab_size - 256
ids = list(tokens) # copy so we don't destroy the original list

merges = {} # (int, int) -> int
for i in range(num_merges):
  stats = get_stats(ids)
  pair = max(stats, key=stats.get)
  idx = 256 + i
  # print(f"merging {pair} into a new token {idx}")
  ids = merge(ids, pair, idx)
  merges[pair] = idx

print("tokens length:", len(tokens))
print("ids length:", len(ids))
print(f"compression ratio: {len(tokens) / len(ids):.2f}X")
```

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
for (p0, p1), idx in merges.items():
    vocab[idx] = vocab[p0] + vocab[p1]

def decode(ids):
  # given ids (list of integers), return Python string
  tokens = b"".join(vocab[idx] for idx in ids)
  text = tokens.decode("utf-8", errors="replace")
  return text

def encode(text):
  # given a string, return list of integers (the tokens)
  tokens = list(text.encode("utf-8"))
  while len(tokens) >= 2:
    stats = get_stats(tokens)
    pair = min(stats, key=lambda p: merges.get(p, float("inf")))
    if pair not in merges:
      break # nothing else can be merged
    idx = merges[pair]
    tokens = merge(tokens, pair, idx)
  return tokens
```
